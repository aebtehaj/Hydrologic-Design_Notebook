# 1- Linear Reservoir and Convolution 

A linear reservoir is a simple and meaningful representation of the watershed response to precipitation events. Continuity equation for a linear reservoir is written as follows:

$$\begin{cases}

\frac{dS(t)}{dt}  = I(t)-Q(t) \quad \text{continuity equation} \\

S(t)  =K\,Q(t) \qquad \quad  \text{linear storage function}

\end{cases}$$


<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F1.png" alt="Figure 1" style="width: auto; height: 200px;">
</div>

<p style="text-align: center;">Schematic of a linear reservoir. A reservoir is linear if we assumed that the storage function $S\left(t\right)$ is linearly related to the outflow through the constant K.</p>

where $Q\left(t\right)$ is the outlfow, $I(t)$ is the inflow $\left\lbrack {\mathrm{m}}^3 {\mathrm{s}}^{-1} \right\rbrack$ (**rainfall excess**), $dS(t)/dt$ is change in the basin storage over time, and $K$ represents a time scale $\left\lbrack \mathrm{s}\right\rbrack$ that captures the \textbf{residence time} of the reservoir. Here, the storage $S(t)$ is the sum the rainfall excess minus the outflow up to time $t$.  

Now upon substitution, we obtain and have to solve a **first-order ordinary differential equation** (ODE):

$$K\frac{dQ(t)}{dt}=I(t)-Q(t)$$

$$\frac{dQ(t)}{dt}=I(t)/K-Q(t)/K$$

Multiplying both sides by $e^{t/K}$, one can get:

$$e^{t/K} \frac{dQ(t)}{dt}+e^{t/K} Q(t)/K=e^{t/k} I(t)/K.$$	

Now because of the product rule, we can rewrite the above equation as follows:

$$\frac{d\left(Q(t)e^{t/K} \right)}{dt}=I(t)\frac{1}{K}e^{t/K}$$

and thus, we have,

$$\int_0^t d\left(Q(s)e^{s/K} \right)=\int_0^t I(s)\,\frac{1}{K}\,e^{s/K} \,ds$$

Therefore, we can obtain the outflow of a linear reservior as follows:

$$Q(t)=Q_0 e^{-t/K} +\int_0^t I(s)\frac{1}{K}e^{-(t-s)/K} ds$$

## 1-1 Concept of Convolution 

The above solution states that the outflow at any time is a linear sum of the contributions by an initial flow (first term) that goes to zero over time and the integral term (second  term) that represents the effects of the reservoir resident time $K$ on the outflow. The integral term is a **convolution integral**. Learning the concept of convolution is crucial for understanding the concept of the **unit hydrograph**.

Mathematical convolution for two generic functions $f(t)$ and $g(t)$ is defined as follows:

$$(f*g)(t)=\int_{-\infty }^{+\infty } f(s)\,g(t-s)\,ds=\int_{-\infty }^{+\infty } f(t-s)\,g(s)\,ds$$

We can see that the convolution of a function $f(t)$ with a kernel $g(s)$ may result in <strong>translation (shift)</strong>  and <strong>dilution</strong> of $f(t)$. We will show later on the the rainfall excess can be represented by $f(t)$ and the kernal function or the unit hydrograph is a characteristic of the watershed. The result of the convolution of excess rainfall with the unit hydrograph will the be watershed outflow.</p>

## 1-2 Instantaneous Unit Hydrograph (IUH) 

Now looking back on the integral in the linear reservoir outflow equation, we can see that the inflow $I(s)$ resembles $f\left(s\right)\;$ and the convolution kernal $u(t-s)=\frac{1}{K}e^{-(t-s)/K}$ resembles $g(t-s)$


$$(I*u)(t)=\int_0^{\infty } I(s)u(t-s)ds  \;\;\;\textrm{and}\;\;\;   u(t)=\frac{1}{K}\,e^{-t/K}$$

 If we define the input rainfall excess as a unit impulse function as follows:


$$I(s)=\begin{cases}
0 & \text{if}\,\, s\neq 0\\
1 & \text{if}\,\, s=0
\end{cases}$$

Then, we can define the **impulse response** function of a linear reservoir as follows:

$$ (I*u)(t) = \int_0^\infty I(s)u(t-s)ds = u(t) =  \frac{1}{K} e^{-t/K}. $$

The function $u(t)=\frac{1}{K}\,e^{-t/K}$ is often called the **instantaneous unit hydrograph (IUH)**, which is the response of a linear reservoir to a **unit excess rainfall** with an **infinitesimally small duration**.

## 1-3 Unit Hydrograph

Similar to the above formalism for a linear reservoir, a watershed has a response function to a **unit pulse** of excess precipitation --- albeit a more complex one. This response function is known as the **unit hydrograph,** which is a **time-invariant characteristic** of a watershed. More specifically, the unit hydrograph $u(t)$ of a watershed is defined as tthe **direct runoff hydrograph (DRH)** resulting from a **unit pulse of excess rainfall**, which is defined as a unit depth of excess rainfall (i.e., 1 $\left\lbrack \textrm{in}\right\rbrack$, 1 $\left\lbrack \textrm{cm}\right\rbrack$, etc) distributed uniformly over the entire watershed for an effective duration of $\triangle t$.

Note that convolution is a linear operator

$$f*\left\lbrack a\,g_1 (t)+b\,g_2 (t)\right\rbrack =a\,f*g_1 (t)+b\,f*g_2 (t),$$

which means that we can **superimpose** the impacts of **two pulses of excess rainfall** over a watershed through **linear combination of their individual outflow responses**.

Therefore, having the **unit hydrograph**, one can obtain the **outflow of a watershed** as convolution of excess rainfall $p_e (t)$ with the unit hydrograph $u(t)$ as follows:


$$ Q(t) = \int_{0}^{t} p_e(s) u(t-s) ds. $$

# 2- Computation of the Unit Hydrograph

<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F6.png" alt="Figure 6" style="width: 500px; height: auto;">
</div>
<p style="text-align: center;"> Convolution of excess rainfall hyetograph (ERH) at $m=1\ldots M$ intervals with the unit hydrograph of a basin, sampled at $l=1\ldots L$ points. The convolution results in a \textbf{Direct Runoff hydrograph (DRH)} at $n=1\ldotsN$ points, where $N=M+L-1$.</p>

Since, hydrologic data is not continuous, we must use a discretized version of the convolution equation as follows:

$$Q_n =\sum_{m=1}^{min(n,M)} p_m u_{n-m+1} ~~~~n=1,...,N\;\textrm{and}\;N=M+L-1.$$


- $\displaystyle Q_n :\;\textrm{Direct}\;\textrm{runoff}\;[L^3 T^{-1} ]\;\textrm{,}\;n=1,...,N$
- $\displaystyle p_m :\;\textrm{Excess}\;\textrm{rainfall}\;\textrm{[L],}\;m=1,...,M$
- $\displaystyle u_{n-m+1} \textrm{:}\;\textrm{Unit}\;\textrm{hydrograph}\;[L^2 T^{-1} ]\textrm{,}\;l=n-m+1\textrm{.}$  


For the example shown in the previous figure, we have

$$Q_n =p_1 \,u_n +p_2 \,u_{n-1} +p_3 \,u_{n-2} ,$$

where $M=3\textrm{,}\;L=6$, $N=M+L-1=8$. Therefore, the above convolution can be expressed as a linear system of equation as follows:

$$\begin{array}{l}
Q_1 =p_1 u_1 +0+0\\
Q_2 =p_2 u_1 +p_1 u_2 +0\\
Q_3 =p_3 u_1 +p_2 u_2 +p_1 u_3 \\
Q_4 =p_3 u_2 +p_2 u_3 +p_1 u_4 \\
Q_5 =p_3 u_3 +p_2 u_4 +p_1 u_5 \\
Q_6 =p_3 u_4 +p_2 u_5 +p_1 u_6 \\
Q_7 =p_3 u_5 +p_2 u_6 +0\\
Q_8 =p_3 u_6 +0+0\\

\end{array}$$


Through convolution, one can obtain the above set of equations through **flipping the rainfall excess hyetograph and slide it over the coordinates of the unit hydrograph and then compute the sum over multiplications.**

$$p_{3\;} |p_2 |p_1 \;\left(\textrm{flipped}\right)\;\;\;\longrightarrow \left(\textrm{multiplication}+\textrm{summation}\right)\;\textrm{with}\;\textrm{zero}\;\textrm{padded}\;\textrm{times}\;\textrm{series}\;\textrm{of}\;\textrm{the}\;\textrm{unit}\;\textrm{hydrograph}\;\;\mathrm{0}|0|u_1 |u_2 |u_3 |u_4 |u_5 |u_6 |0|0$$   

## 2-1 Forward substitution and least-squares

We can solve the above set of equations through forward substitution. In other words, one can obtain $u_1 =\frac{Q_1 }{p_1 }$ and then use the second equation to obtain $u_2$ and so forth. 

One can arrange all data of the above linear system of equations in a **matrix form** as follows:


<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F7.png" alt="Figure 7" style="width: 500px; height: auto;">
</div>

and solve it for $u_i$ values to obtain the **unit hydrograph** from the **observed** direct **runoff hydrograph** of a precipitation excess event. 

We can see that in the above linear system of equations we have **more equations that unknown**. Therefore, calculation of the unit hydrograph is often an **overdetermined problem**.

The above system of linear equation, which is resulted from a discrete convolution operation can be rewritten in matrix form as follows:

$$\mathbf{q}=[Q_{1},Q_{2},\ldots,Q_{N}]^{T} , \mathbf{u}=[u_{1},u_{2},\ldots,u_{L}]^{T}, L=N-M+1.$$

<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F8.png" alt="Figure 8" style="width: 700px; height: auto;">
</div>

where $q$ is an $N\textrm{-}\textrm{by}\textrm{-}1$ column vector, $P$ is an $N\textrm{-}\textrm{by}\textrm{-}L$ matrix, and $u$ is an $L\textrm{-}\textrm{by}\textrm{-}\textrm{1}$ column vector. 

For the provided example, the matrix form is as follows:


$$ \mathbf{P}\mathbf{u} = \mathbf{q} $$


<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F9.png" alt="Figure 9" style="width: 700px; height: auto;">
</div>

The matrix form can be easily implemented into a software library for numerical linear algebra. As previously explained, the equation **$Pu=q$** is an **over-determined linear system**, since we have more equations than unknowns. Two of the main methods used to solve this system of equations for $u$ are: 

1. **Forward/backward substitution:** Write out the first or last $L$ equations. Solve the first or last equations and sequentially substitute and solve the remaining equations. In this formalism only $L$ equations are used, which result in multiple solutions to the problem. 

2. **Least squares:** Using linear algebra, we can derive a formula to calculate the least squares estimate of the unit hydrograph ordinates as follows:


$$ \mathbf{Pu}  =\mathbf{q}\\
\mathbf{P}^{T}\mathbf{Pu}  =\mathbf{P}^{T}\mathbf{q}\\
\mathbf{u} =\left(\mathbf{P}^{T}\mathbf{P}\right)^{-1}\mathbf{P}^{T}\mathbf{q} $$

where $(.)^T \;$ is the matrix transposition operator,  which flips the elements of a matrix over its diagonal elements. Note that by least squares, we mean that the solution has minimum error variance amoung all other possible solutions and the solution is unique. 

The above matrix formulation of the linear system can be solved efficiently using a technical programing language such as MATLAB. Note that, in the above formulation, we multiply both sides with $P^T$ to obtain a square matrix ${{PP}}^T$, which might be invertible. We need to note that $P\;$ is often a rectangular matrix with more rows than columns ($N\gg L$) and cannot be inverted. 

The matrix ${\left(P^T P\right)}^{-1} P^T$ is called **pseudo inverse** of $P$. In Pyhton, computation of the unit hydrograph can be simply implelmented using either of the following commands: $\textrm{np.linalg.solve(P, q)}$ or $\textrm{np.dot(np.linalg.pinv(P), q)}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import toeplitz

# Time array
t = np.arange(0, 5.5, 0.5)

# Direct runoff hydrograph (DRH)
q = np.array([423, 1923, 5297, 9131, 10625, 7834, 3921, 1846, 1402, 830, 313])

# Excess rainfall hyetograph (ERH)
p = np.array([1.06, 1.93, 1.81, 0, 0, 0, 0, 0, 0, 0, 0])
p_1 = np.array([1.06, 1.93 - 0.5, 1.81 + 0.2, 2, 0, 0, 0, 0, 0, 0, 0])

r = np.array([1.06, 0, 0, 0, 0, 0, 0, 0, 0])
P = toeplitz(p_1, r)

# Unit Hydrograph
# u = np.linalg.inv(P.T @ P) @ P.T @ q
# or
u = np.linalg.pinv(P) @ q

# Plotting
fig, ax1 = plt.subplots(figsize=(15, 8))

plt.rcParams["font.size"] = 22
plt.rcParams["font.family"] = "Times New Roman"

ax1.set_xlabel('Time (hr)')
ax1.set_ylabel('Discharge (cfs)')
ax1.plot(t, q, label='DRH', linewidth=2)
ax1.plot(t[:len(u)], u, label='UH', linewidth=2)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.set_ylabel('Precipitation excess (in)')
ax2.bar(t, p, alpha=0.3, label='Precipitation')

fig.tight_layout()
ax2.legend(loc='upper right')
plt.show()
```

## 2-2 S-Hydrograph

As we mentioned, the UH is a **time invariant** characteristic of a watershed in response to a **unit pulse** of **excess rainfall** with a **specific duration** $\Delta t$.  **However, since the UH is obtained through a linear process, we can change a unit hydrograph from one duration to another.** To that end, we need to derive the **S-hydrograph** based on the principle of superposition. 

By definition, the S-hydrograph is the direct runoff hydrograph of a watershed resulting from a continuous excess rainfall at a constant rate of 1 [cm/hr] or [in/hr] for an infinite period of time.

In other words, the S-hydrograph $g(t)=\Delta t\,[u(t)+u(t-\Delta t)+u(t-2\Delta t)+\ldots]$ is the response of the watershed to summation of an infinite number of unit pulse of excess rainfall with intensity $\frac{1}{\Delta t}$ and duration $\Delta t$  -- or a unit rainfall excess over an infinite period of time (Figure below). 

<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F11.png" alt="Figure 11" style="width: 1000px; height: auto;">
</div>

<p style="text-align: center;">Conceptual representation of the S-hydrograph</p>

As a result, the unit hydrograph for $\Delta t^{\prime } =n\times \Delta t$ duration, **with integer** $n$ **values**, can be obtained as follows:

$$ u\left(\Delta t^{\prime}\right)=\frac{1}{\Delta t^{\prime}}\left[g(t)-g(t-\Delta t^{\prime})\right] $$


<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F12.png" alt="Figure 12" style="width: 1000px; height: auto;">
</div>

<p style="text-align: center;">Derivation of the unit hydrograph with duration $\Delta t^{\prime } =n\Delta t$, where $n\in \mathbb{N}$, from an S-hydrograph that is generated from a unit hydrograph with duration $\Delta t$.</p>

**Example Problem 7.3:** Use 0.5-hr unit hydrograph in the following table to calculate 1.5-hr unit hydrograph. 


**Solutions**:

<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F13.png" alt="Figure 13" style="width: 1000px; height: auto;">

A shifted version of the hydrogrsph is shown between columns 2 and 3. Column 3 represents the sum of the unit hydrographs multiplied by $\Delta t=0.5$ [hr]. Column 4 shows the lagged unit hydrograph for $\Delta t=1.5$ [hr]


An excel file solving an example to derive a 1.5-h unit hydrograph from 0.5-hr unit hydrograph is posted in the course website. Column 5 computes the unit hydrograph for $\Delta t^{\prime } =1.5\;$ [hr].

```python
import numpy as np

# Define time intervals
dt = 0.5  # [hr]
dt_prime = 1.5  # [hr]
t = np.arange(0.5, 6.1, dt)
u_0_5 = np.array([404, 1079, 2343, 2506, 1460, 453, 381, 274, 173, 0, 0, 0])

# Initialize the matrix M with zeros
M = np.zeros((len(t), len(t)))

# Fill the matrix M with lagged values of u_0_5
for i in range(len(t)):
    u_temp = u_0_5[:len(u_0_5)-i]
    M[i:len(t), i] = u_temp

# Calculate the S-hydrograph
s_hydo = dt * np.sum(M, axis=1)

# Lag the S-hydrograph
s_hydro_lag = np.concatenate([np.zeros(3), s_hydo[:len(s_hydo)-3]])

# Compute the 1.5-hour hydrograph
u_1_5 = (1 / dt_prime) * (s_hydo - s_hydro_lag)

# Print the results
print('1.5-hour hydrograph:')
print(u_1_5)
```

## 2-3 Synthetic Unit Hydrographs

Thus far, the methods we have described only apply for the basin with gage measurements of the streamflow, however, many basins are **ungauged**.  Therefore, engineers have analyzed many unit hydrographs from gauged basins to obtain a universal functional representation for the unit hydrograph that can be generalized to ungauges basins. This representation of the unit hydrograph is often called **Synthetic Unit Hydrograph (SUH)** that can be applied to **ungauged basins**. These synthetic methods typically scale a general UH shape or calculate key ordinates based on easily determined characteristics of watershed and excess precipitation. 

There are many of these methods, such as Snyder's or Clark's SUH; however, we will cover the very popular **NRCS (SCS)** dimensionless synthetic hydrograph method. The dimensionless SCS hydrograph and its triangular representation are shown in Figure below and tabulated.

**Input parameters:**

- $t_R$: duration of excess rainfall 
- $t_L =0.6t_c$\textbf{: time lag between the centroid of rainfall excess hyetograph to the peak discharge}
- $t_c$: \textbf{time to concentration}
- $t_p =\frac{t_R }{2}+t_L$: **time to peak**
- $t_r =t_b -t_p$: recession time
- $t_b$: base time for total duration of the scaled SUH $t_b =2\ldotp 67t_p \;$ (triangle) and $t_b =5t_p$ (curvilinear) 
- $q_p$: **peak flow of the scaled SUH**.

### 2-3-1 Concentration time

As we previously discussed the time of concentration ($t_c$) for a watershed is the time for a raindrop to travel from the farthest point in the watershed to the outlet. The SCS recommends two methods for calculation of $t_c$: 

**The lag method:** defines the time lag as the time in hours from the center of the mass of the rainfall excess to the peak discharge, which can be computed as follows: 

$$t_L =\frac{L^{0.8} (S+1)^{0.7} }{1900Y^{0.5} }$$


where $Y\;$ is the average slope of the watershed in \%, $L$ denotes the hydraulic length of the main channel in feet, and $S=\frac{1000}{\textrm{CN}}-10$  [**inches**] represents the potential maximum retention that is obtained from the curve numbers (CN). The SCS method suggests the following relationship for the concentration time.

$$ t_c = \frac {5}{3}\,t_L  \qquad \text{[hr]} $$

**The upland or velocity method:** defines the concentration time as follows: 


$$ t_c = \frac{L}{3600V} \qquad \text{[hr]} $$

where $L\;$ and $V$ are hydraulic length and velocity in [ft] and [ft/s], respectively. The velocity of the upland method can be obtained from Figure 8 based on the average slope of the watershed. 

Clearly, if we discretize the the stream into $k$ segments, the concentration time is the sum of travel time for different stream segments as we discussed before,

$$ t_c= \frac{1}{3600}\Sigma_{i=1}^{k} \frac{L_i}{V_i} \qquad \text{[hr]} $$

### 2-3-2 Time to Peak

Time to peak ($t_p$)  is the time from the beginning of the rainfall to the time of the peak discharge, which can be expressed as a function of the time lag $t_L$ as follows:

$$t_p =\frac{t_R }{2}+t_L$$

where $t_p$ is in $\left\lbrack \textrm{hr}\right\rbrack$, $t_R$ is the duration of the rainfall excess in $\left\lbrack \textrm{hr}\right\rbrack$, and $t_L$ is the lag time in $\left\lbrack \textrm{hr}\right\rbrack$. 

When $t_R$ is not known, the SCS method suggests

$$t_R =0.133t_c$$

and because $t_L =\frac{3}{5}\,t_c =0.6t_c$, one can obtain the following relationship for approximating the time to peak:

$$t_p =\frac{0.133t_c }{2}+0.6t_c =0.67t_c$$


### 2-3-3 Peak Discharge

From continuity equation, we know that the area under the unit hydrograph is equal to the volume of the excess unit rainfall - example 1 [in] all over the watershed. Here, the unit of $q_p$ in SCS method is the unit of discharge per depth of rainfall excess such as $\left\lbrack {\mathrm{m}}^3 {\mathrm{s}}^{-1} {\textrm{cm}}^{-1} \right\rbrack$ or $\left\lbrack \textrm{cfs}\;{\textrm{in}}^{-1} \right\rbrack$.

Therefore, base on conservation of mass of the depth of the rainfall and the triangular shape of the proposed SCS unit hydrograph, we have

$$1=\frac{1}{2}\frac{q_p }{A}(t_p +t_r ).$$

which can be rearranged as

$$q_p =\left(\frac{1}{1+t_r /t_p }\right)\frac{2A}{t_p }$$

For the SI unit, in the above equation the area should be in  $\left\lbrack {\mathrm{m}}^2 \right\rbrack$ and time is in second to obtain the discharge in $\left\lbrack {\mathrm{m}}^3 {\mathrm{s}}^{-1} \right\rbrack$. To convert the above formula to be used for watershed area in square kilometers and hourly time scale, we need to to multiply the above equation by $\frac{{10}^4 }{3600}$.

---

The SCR method suggests $t_r =1.67t_p$ and thus 

$$q_p =\frac{c\,A}{t_p }$$


where $c=\left(\frac{\;1}{1+1\ldotp 67}\right)\times 2\times \frac{{10}^4 }{3600}=2\ldotp 08$ when area is in square kilometers. The $c\;$ value in English unit is 483.4, where $A$ is in square miles.

# 4- Flow Routing

**Flow routing** is the procedure of determining the downstream hydrograph from hydrographs at one or more points upstream. 

As water enters a channel, either from hillslopes or upstream channels, the flow velocity **shifts** the location of the peak flow and surface roughness and storages along the water path **attenuate** the magnitude of the peak.  The flow routing is referred to a class of methods that calculates the **redistribution** (widening) and **translation** of upstream hydrograph as a function watershed flow characteristics.

<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F19.png" alt="Figure 19" style="width: 800px; height: auto;">
</div>

<p style="text-align: center;"> Left: Schematic of redistribution and translation (Chow et al. 1988). Middle: The storage accumulation and release during the flow routing. Right: Conceptualization of channels as storage reservoirs (COMET Program) </p>

There are two main types of routing methods:

1. **Lumped (Hydrologic) Routing**: Flow is calculated at a particular point as a function of time alone.  
2. **Distributed (Hydraulic) Routing**: Flow is calculated as a function of time and space.

For this course, we focus on the simpler lumped modeling.  

Within lumped modeling, there are multiple methods depending on what system we are trying to model.  We are going to focus on two common methods for flow routing in reservoirs (**level-pool method**) and stream channels (**Muskingum method**).

Derivation of routing methods begins with the continuity equation:

$$\frac{dS(t)}{dt}=I(t)-Q(t)$$

In routing,$I(t)$ is the known inflow hydrograph, while $S(t)$ and $Q(t)$ are unknown. Just as in our discussion of linear reservoirs, we need a **storage-discharge** relationship, $S=f(Q)$. 

There are two types of storage-discharge relationships: 

- (1) **invariable reservoirs**:  The invariable are those in which there is a one-to-one relationship between the outflow and the reservoir storage. This scenario is valid in short and wide reserviors. 
- (2) **variable reservoirs**: In variable reservoirs, the relationship between the outflow and storage capacity is not unique and follows a loop when the reservoir is filled or emptied. This nonlinear relationship mostly occurs in long and narrow reservoirs where the water velocity is significant and produces **backwater effects** in the reservoir.

## 4-1 Level Pool Routing 

Level pool routing typically applies to man-made hydraulic structures like reservoirs, dams and stormwater ponds that are common in many watersheds. **The key assumption is a horizontal water surface and negligible velocity in the reservoir**. This allows an invariable relationship between storage (water surface height) and discharge, because any change in storage results in a uniform change in height of water surface.  

Beginning with the continuity equation $\frac{dS}{dt}=I(t)-Q(t)$ , we have $dS=I(t)dt-Q(t)dt$, one can discretize it as follows assuming that the changes in inflow and outflow are linear over a small timestep $\triangle t$:

$$S_{j+1} -S_j =\frac{I_j +I_{j+1} }{2}\triangle t-\frac{Q_j +Q_{j+1} }{2}\triangle t$$

<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F22.png" alt="Figure 22" style="width: 1000px; height: auto;">
</div>

<p style="text-align: center;"> Discretization of the reservoir continuity equation over a small time interval </p>

The inflow $I_j$ and $I_{j+1}$  are known at all times while initial storage $S_j$ and outflow $Q_j$ are specified as initial conditions, so we can re-arrange the equation into knowns and unknowns as follows:

$$\left\lbrack \frac{2S_{j+1} }{\triangle t}+Q_{j+1} \right\rbrack =I_j +I_{j+1} +\left\lbrack \frac{2S_j }{\triangle t}-Q_j \right\rbrack ~~~~(1)$$

To use this equation, we need the **elevation-storage function** $S=f(H)$ and **elevation-discharge** $Q=f(H)$ relationships of the reservoir to create a  **storage-outflow function** that allows obtaining the outflow at time step $j+1$ from the left-hand side of the above equation $\frac{2S_{j+1} }{\Delta t}+Q_{j+1}$ iteratively. 

- The **elevation-storage function** $S=f(H)$ is simply constructed from the geometry of the reservoir or pond.
- The **elevation-discharge** $Q=f(H)$ is slightly more difficult to obtain; however, the outflow of storage facilities is often controlled by hydraulic structures such as sluice gates, weirs, pipes, and orifices which typically have well-defined equations based on thr water head (Figure above). 

For example, for an uncontrolled ogee crest spillway, the outflow as a function of total water head on the crest is:

$$Q=C_0 LH^{3/2}$$

$L:$ is the effective length of the crest
$C_0 :$ is the discharge coefficient 
$Q:$ is discharge in [cfs].

## 4-2 Hydrologic River Routing 

In order to calculate the routing for streams in a watershed, **we can no longer assume that the water surface is level**, since there is not negligible velocity in a stream and there are backwater effects. Therefore, we need to define a different storage-outflow equation, $S=f(Q)$. 

The Muskingum method assumes that storage in a channel is the sum of a prism and a wedge storage.  

- The **prism storage** is a linear reservoir defined as $S=KQ$ where $K$ is a constant representing approximately the travel time through the reach.  
- The **wedge storage** is defined as $S=KX(I-Q)$, where $X$ is a weighting factor ranging from 0 to 0.5 and thus,

$$S=KQ+KX(I-Q)=K(XI+(1-X)Q)$$

This **wedge storage** represents a **flood wave** as it propagates through a stream channel as shown in Figure below.

<div style="text-align: center;">
  <img src="Figures/Ch7/Ch7F27.png" alt="Figure 27" style="width: 700px; height: auto;">
</div>

<p style="text-align: center;"> Schematics of wedge storage and flood wave propagation assumed in Muskingum method (Credit: The COMET Program) </p>

We can discretize this storage equation as follows for time step $jth$ 

$$S_j =K[XI_j +(1-X)Q_j ]~~~~S_{j+1} =K[XI_{j+1} +(1-X)Q_{j+1} )]$$

and thus:

$$S_{j+1} -S_j =K\left\lbrace \left\lbrack (XI_{j+1} +(1-X)Q_{j+1} \right\rbrack -\left\lbrack XI_j +(1-X)Q_j \right\rbrack \right\rbrace ~~~~(2)$$

Recall, we previously defined the change in storage as follows:

$$S_{j+1} -S_j =\frac{I_j +I_{j+1} }{2}\triangle t-\frac{Q_j +Q_{j+1} }{2}\triangle t~~~~(3)$$

Combining equations 2 and 3, one can obtain the following flood routing equation:, known as the Muskingum stream routing formulation:

$$Q_{j+1} =C_1 I_{j+1} +C_2 I_j +C_3 Q_j$$

$$C_1 =\frac{\triangle t-2KX}{2K(1-X)+\triangle t}~~~~C_2 =\frac{\triangle t+2KX}{2K(1-X)+\triangle t}~~~~C_3 =\frac{2K(1-X)-\triangle t}{2K(1-X)+\triangle t}$$

where $C_1 +C_2 +C_3 =1$. 

Typically, the streams is broken up into $N$ subreaches when performing this calculation.  The U.S. Army Corps of Engineers recommends the following condition to ensure solution stability when we discretize the river reach:

$$\frac{1}{2(1/X)}\le \frac{K}{N\triangle t}\le \frac{1}{2X}$$
