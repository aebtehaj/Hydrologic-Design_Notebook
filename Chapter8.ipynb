# 1- Introduction

Hydraulic engineering studies physical properties of a fluid flow in pipes and/or open channels. Pipe flow is typically a pressurized flow while an open channel flow has free water surfaces and is under atmopsheric pressure. 

Volumetric flux of pipe and open channel flows or the flow discharge can be calculated as follows:

$$Q=\int_A vdA$$

where $v$ is the flow velocity $\left\lbrack \mathrm{L}{\mathrm{T}}^{-1} \right\rbrack$, and $\mathrm{A}\left\lbrack {\mathrm{L}}^2 \right\rbrack$ represents the cross sectional area. We represent the flow discharge through the above integral equation because in reality the flow velocity is **not uniformly distributed** over the cross section of the channel and is not constant.

Obviously, the commonly used $Q=\textrm{VA}$ is an approximation of the flow discharge, where$\mathrm{V}$represents the **average velocity**. We will use some correction factors to correct for the computation of kinetic energy of the flow, when the effects of the non-uniform velocity might be significant.

# 2- Energy and Momentum in Open Channel Flows
## 2-1- Conservation of Energy 

As we discussed before, there are three sources of energy that move the water parcels in a channel:

- Potential energy: $\textrm{mgz}$ 
- Kinetic energy: $\frac{1}{2}mV^2$ 
- Pressure energy: $p\forall =p\frac{m}{\rho }$ (Here, $\forall$ denotes the volume) 

$$E=mgz+\frac{1}{2}mV^2 +p\frac{m}{\rho }$$ 

We often express the above energy per unit weight of the water parcel as follows:

$$\frac{E}{mg}=z+\frac{V^2 }{2g}+\frac{p}{\gamma }$$

**Hydraulic Grade Line (HGL) and Energy Grade Line (EGL)**

Hydraulic Grade Line (HGL) =  $z+\frac{p}{\gamma }$

Energy Grade Line (EGL) =  $z+\frac{V^2 }{2g}+\frac{p}{\gamma }$

$z$: elevation head $\left\lbrack \mathrm{L}\right\rbrack$   $\frac{p}{\gamma }$: pressure head $\left\lbrack \mathrm{L}\right\rbrack$   $\frac{V^2 }{2g}$: velocityh head $\left\lbrack \mathrm{L}\right\rbrack$


There are always some **energy dissipation** due to **viscous stress and friction**. These energy losses per unit weight of water are expressed as the **head loss** $h_L$. 

Therefore, the energy equation for two points on the flow path can be written as follows:

$$z_1 +\frac{p_1 }{\gamma }+\frac{V_1^2 }{2g}-{h_L }=z_2 +\frac{p_2 }{\gamma }+\frac{V_2^2 }{2g}$$

As we discussed, in reality, the velocity is not uniform across the flow cross section. To correct for this non-uniformity in the energy equation, we need to account for the variability of the kinetic energy. To that end, we need to recall that the **mass of the fluid that flows through a differential area** $\textrm{dA}$ **per unit time is** $\rho \textrm{vdA}$.

Therefore, the total kinetic energy of the flow per unit time is as follows:

$$\frac{1}{2}(\rho vdA)v^2 =\frac{1}{2}\rho v^3 dA$$

Thus, we can obtain a velocity correction factor as follows:

$$\alpha \frac{1}{2}\rho V^3 A=\int_A \frac{1}{2}\rho v^3 dA$$

$$\alpha =\frac{1}{AV^3 }\int_A v^3 dA$$

where $\mathrm{v}$ is the fluid velocity, which varies across the cross section, and $V$ denotes the **mean velocity**.

Therefore, for a an open channel flow with non-uniform velocity profile we have,

$$z_1 +\frac{p_1 }{\gamma }+\alpha \frac{V_1^2 }{2g}-h_L =z_2 +\frac{p_2 }{\gamma }+\alpha \frac{V_2^2 }{2g}$$

In pipes $\alpha =$ 2 for laminar flows and $\alpha =$ 1.03 - 1.06 for turbulent flows. In open channels, $\alpha$ ranges between 1.10 and 1.20 and varies from 1.5 to 2.0 in river flows.

**Head Loss in Pipe Flows:**

From the Darcy–Weisbach equation, we have,

$$h_L =\frac{fL}{D}\frac{V_2^2 }{2g},$$

where $h_L$ is the head loss due to pipe friction, $f$ is the dimensionless friction factor, $L$ is the length of the conduit, $D$ is the diameter, $V$ is the mean flow velocity.


In **laminar pipe flows** ($\textrm{Re}<2000$) , $f=\frac{64}{R_e }$, where the Reynold’s number is $R_e =\frac{\textrm{vD}}{\nu }$.

In **turbulent pipe flows** ($\textrm{Re}>2000$), the roughness factor $f$ depends on the relative roughness $\frac{k_s }{D}$ and Reynolds number, where $k_s$ is the average of the roughness height of the pipe internal surface. 


In turbulent flows, there is the region in the vicinity of the boundary in which the flow is laminar, this layer is called the **viscous sublayer** or the **laminar sublayer**.

Depending on the thickness of the viscous sublayer and the height of the surface roughness elements, the boundary layer flow can be categorized as a **smooth** or **rough flow**.

- If the thickness of the viscous sublayer is much greater than the size of the roughness elements, the flow is **dynamically smooth**.
- If the height of the rough elements are larger than the thickness of the viscous sublayer, the flow is **dynamically rough**. In this case the flow between the roughness elements is still turbulent and pressure forces dominate the viscous forces.

The quantity $\frac{\nu }{u_{\;*} }$ is called the viscous length scale, which is proportional to the thickness of the viscous sublayer, where $\nu$ is the kinematic viscosity and $u_{\;*} =\sqrt{\frac{\tau_0 }{\rho }}$ is the shear velocity.

**Head Loss in Uniform Open Channel Flow:**

The friction slope in a flow is as follows

$$h_L =z_2 -z_1 \to \frac{h_L }{L}=\frac{z_2 -z_1 }{L}=S_f$$

In a **uniform flow** the spatial flow properties, $y_1 =y_2$ and $\;V_1 =V_2$, do not change. Thus, the friction slope ($S_f$), the water surface ($S_w$) and channel slope ($S_0$) are equal. 

$$\frac{z_2 -z_1 }{L}=S_f =S_0 =S_w$$

**Momentum Equation in Open Channel Flow:**

For a fluid control volume of a uniform or non-uniform flow, we have the conservation of momentum as follows:

$$\sum F=\frac{d(mv)}{dt}=\sum_i (\rho V_i .A_i )*V_i \;\;\; \textrm{Newton's second law} $$

For the uniform flow, we get

Downslope gravitational force: $F_g =\gamma \textrm{ALsin}\theta$

Friction force: $F_f =\tau_0 \textrm{PL}$

where $\tau_0$ is the shear stress at the bottom of the channel, $P$ is the wetter perimeter of the channel, $L$ is the channel length, $\gamma$ denotes the specific weight of the fluid, $A$ is the wetted area of the flow cross section, and $\theta$ is the slope of the channel bottom. When the channel is gently sloped (i.e., $\theta$ is small), we have $\sin \theta =\tan \theta =S_0 =S_f$ .

As a result, one can obtain $\tau_0 \textrm{PL}=\gamma \textrm{AL}S_{f\;} =\gamma \textrm{AL}S_{0\;}$ and thus

In a **fully turbulent flow**, the shear stress is proportional to the square of the mean flow velocity as follows:

$$\tau_0 =C_f \rho V^2 =\gamma RS_0$$

where $C_f$ is a dimensionless proportionality constant. As a result, one can obtain,


$$V=\sqrt{\frac{g}{C_f }}\sqrt{RS_0 }=C\sqrt{RS_0 }$$


where $C=\sqrt{\frac{g}{C_f }}$  is called the **Chezy coefficient**

Clearly, in a uniform flow, one can relate the Chezy coefficient to the Darcy-Weisbach roughness coefficient as follows:

$$\frac{h_L }{L}=S_f =S_0 =\frac{f}{4R}\frac{V^2 }{2g}\to V=\sqrt{\frac{8g}{f}}\sqrt{RS_0 }\to C=\sqrt{\frac{8g}{f}}$$

Robert Manning (1895) proposed the following experimental formula for the Chezy coefficient:

$$C=\frac{1}{n}R^{\frac{1}{6}} \;\; \textrm{(SI)} $$

where $n$ is called the Manning roughness coefficient. Therefore, we have

$$V=\frac{1}{n}R^{\frac{2}{3}} S_0^{\frac{1}{2}} \;\; \textrm{(SI)} $$
$$V=\frac{1.49}{n}R^{\frac{2}{3}} S_0^{\frac{1}{2}} \;\; \textrm{(EN)} $$


Manning equation is valid for a fully developed turbulent channel flow, where

$$ n^6 \sqrt{RS_f }\ge 1.1*10^{-13} \;\; \textrm{(SI)\;\; or\;\;} n^6 \sqrt{RS_f }\ge 1.9*10^{-13} \;\; \textrm{(EN)} $$

**Equivalent Roughness:**

Open channel can be made with different materials with different roughness properties. The question is–**how can we obtain an average roughness coefficient that represents well the flow roughness properties?** There are multiple methods. Here, we briefly cover the method by Horton and Einstein.

<div style="text-align: center;">
  <img src="Figures/Ch8/Ch8F8.png" alt="Figure 8" style="width: auto; height: 300px;">
</div>

<p style="text-align: center;">Channel cross section with different wetted perimeters ($P_i$) and roughness coefficients ($n_i$).
</p>

$$V=\frac{1}{n}R^{2/3} S_0^{1/2}$$

$$S_0^{1/2} =\frac{V_1 n_1 }{R_1^{2/3} }=\frac{V_2 n_2 }{R_2^{2/3} }=...=\frac{V_2 n_2 }{R_2^{2/3} }$$

where $V_i$ **is the average flow velocity over** $A_i$, $V$ **is the mean velocity across the entire channel** and $n_e$ **denotes the equivalent roughness**. Given that $R_i =\frac{A_{i\;} }{P_{i\;} }$ and assuming that the flow velocity does not vary significantly across the channel (i.e., $V_{i\;} \sim V$), we have

$$\frac{V_i n_i }{R_i^{2/3} }=\frac{Vn_e }{R^{2/3} }\to (\frac{A_i }{A})^{2/3} =\frac{n_i P_i^{2/3} }{n_e P^{2/3} }\Rightarrow \frac{A_i }{A}=\frac{n_i^{3/2} P_i }{n_e^{3/2} P}$$

Thus, one can obtain,

$$A=\sum A_i =A\frac{\sum n_i^{2/3} P_i }{n_e^{3/2} P}\to n_e =\frac{(\sum n_i^{3/2} P_i )^{2/3} }{P^{2/3} }$$


## 2-2- Best Hydraulic Section:

For a channel with **a constant area and slope**, the best section is the one with the **maximum ability in flow transportation**. This cross section shall have the one with **minimum wetted perimeter** because,

$$Q=\frac{1}{n}AR^{2/3} S_0^{1/2} =\frac{1}{n}\frac{A^{5/3} }{P^{2/3} }S_0^{1/2}\;\;\;\;P_{min} \to Q_{max}$$

<p style="text-align: center;">rectangular (left) and trapezoidal(right) cross sections</p>



**Example**: In a rectangular channel, we have

$P=b+2y=\frac{A}{y}+2y$. Setting $\frac{\partial }{\partial y}P=0$, one can get $-\frac{A}{y^{2\;} }+2=0$, which result in $y=\frac{b}{2}$.

In a trapezoidal cross section, we have $A=\left(b+\textrm{zy}\right)*y$ and thus $b=\frac{A}{\;y}-\textrm{zy}$.

$P=b+2y\sqrt{1+z^2 }=\frac{A}{y}-\textrm{zy}+2y\sqrt{1+z^2 }$. Now $P$ is only as a function of $y$ and $z$.

$$ \begin{cases}
          \frac{\partial }{\partial y}P=0 & \Rightarrow & \frac{A}{\;y^2 }-z+2\sqrt{1+z^2 }=0\;\;\&\;A=\left(b+\textrm{zy}\right)y\Rightarrow b+2\textrm{yz}=2y\sqrt{1+z^2 } \\
          \frac{\partial }{\partial z}P=0 & \Rightarrow & z=\frac{\sqrt{3}}{3} 
\end{cases} $$

and thus $b=\frac{2y}{\sqrt{3}}=\frac{2\sqrt{3}y}{3},R=\frac{y}{\;2}$ and $A=\sqrt{3}y^2$.

- $Q=\frac{1}{\;n}{\textrm{AR}}^{\frac{2}{3}} S_0^{\;\frac{1}{2\;}} =\frac{1}{\;0\ldotp 025}\left(1\ldotp 732y^2 \right){\left(\frac{y}{2}\right)}^{\frac{2}{3}} 0\ldotp {00052}^{\frac{1}{2}} =10$, solving this equation for$y$, we get $y=2\ldotp 38\;\left\lbrack m\right\rbrack$ and $b=2\ldotp 75\;\left\lbrack m\right\rbrack$ and $A=9\ldotp 81\;\left\lbrack m^2 \right\rbrack$ 



## 2-3- Specific Energy:

Assuming $\alpha =1$ and hydrostatic pressure, the total head in open channel flow is as follows:

$$EGL=z+y+\frac{V^2 }{2g}$$

Using the channel bottom as the datum ($z=0$), then we define the **specific energy** as

$$E=y+\frac{V^2 }{2g}=y+\frac{Q^2 }{2gA^2 }$$

The minimum point of this curve determines two important flow regimes.

$$\frac{dE}{dy}=1-\frac{Q^2 }{gA^3 }\frac{dA}{dy}=0$$


Using $T=\frac{\textrm{dA}}{\textrm{dy}}$ as the top width of the flow, one can obtain 

$$\frac{Q^2 }{g}=\frac{A^3 }{T}\to \frac{V^2 }{g}=\frac{A}{T}$$

Defining a new parameter called the hydraulic depth $D=\frac{A}{T}$ , we can conclude that at the minimum specific energy, we have $V=\sqrt{\textrm{gD}}$.

In other words, when the **specific energy is minimum**, the Froude number is $\textrm{Fr}=\frac{V}{\;\sqrt{\textrm{gD}}}$.

<div style="text-align: center;">
  <img src="Figures/Ch8/Ch8F10.png" alt="Figure 10" style="width: auto; height: 500px;">
</div>

Specific energy versus depth for constant discharge. From the equation of specific energy, we can see that as $y\to 0$ then $E\to \infty$ and as $y\to \infty$ then $E\to y$.

We can see that for a given specific energy there are two alternate depths, one for the supercritical and one for the critical flow regime. From fluid mechanics we know that


$$ Fr=\frac{V}{\sqrt{gD}} 
\begin{cases}
          <1 & \textrm{subcritical flow}\\
          =1& \textrm{critical flow} \\ >1 & \textrm{supercritical flow}
\end{cases} $$

The depth associated with the minimum specific energy is called the critical depth.

For a rectangular channel with a critical flow, we have $V=\sqrt{\textrm{gD}}=\sqrt{{\textrm{gy}}_{c\;} }$. Thus, given the flow discharge per unit width of the channel (q), we have 

$$\frac{q}{y_c }=V_c =\sqrt{gy_c }\Rightarrow y_c ={\left(\frac{q^2 }{g}\right)}^{1/3}$$



**Example:** Determine the flow regime and the alternate depth in a rectangular open channel flow when $Q=20\;\left\lbrack m^3 \;s^{\;-1} \right\rbrack$, $b=10\;\left\lbrack m\right\rbrack$ and $y=0\ldotp 6\;\left\lbrack m\right\rbrack$.

**Solution:**

- $\displaystyle V=\frac{Q}{\;\textrm{by}}=3\ldotp 33\;\left\lbrack m^3 s^{-1} \right\rbrack$
- $\textrm{Fr}=\frac{V}{\;\sqrt{\textrm{gy}}}=1\ldotp 37$ and thus the flow is supercritical with $y_c ={\left(\frac{q^{2\;} }{\;g}\right)}^{\;\frac{1}{3}} =0\ldotp 74\;\left\lbrack m\right\rbrack$. 
- $\displaystyle E=y+\frac{Q^2 }{2{\textrm{gA}}^2 }=1\ldotp 166\;\left\lbrack m\right\rbrack \ldotp$ 
- To compute the alternate depth we use the equation of energy for a rectangular channel, which is $E=y+\frac{q}{{\textrm{gy}}^{2\;} }$ . Therefore, we have $1\ldotp 166=y+\frac{2}{\;{\textrm{gy}}^2 }$ and through a root finding algorithm (e.g., bisection method), one can obtain $y_2 =0\ldotp 93\;\left\lbrack m\right\rbrack$ for the subcritical flow.

**Effects of Channel Bottom Humps:**

When we have a hump at the bottom of the channel the flow regime might be changed. In this case typically we study two scenarios:

(1) $E_2 =E_h =E_1 -\Delta z\ge E_{min}$

In this case explanation of the changes in flow velocity and depth is straightforward. Basically, writing the energy equation for a point right upstream of the hump and a point on the hump, we have $E_1 =E_2 +\Delta z$. As a result, the energy over the hump is reduced by $\Delta z$  and thus when the upstream flow is subcritical, we can see from the curve of the specific energy that the flow depth over the hump decreases and velocity increases. On the other hand, when the flow is supercritical the depth over the hump increases and velocity decreases.

(2) $E_h =E_1 -\Delta z<E_{min}$

In this case, the upstream flow needs to adjust its profile to satisfy the conservation of energy. Explanation of the flow profile is possible for a subcritical upstream flow through the specific energy equation; however, when the upstream flow is supercritical, explanation of the flow profile is beyond the scope of this class.

As shown in Figure below, for subcritical upstream flow with depth $y_1$, right upstream of the hump, the energy of the flow increases from $E_1$ to $E_2$ such that $E_2 =E_{\min } +\Delta z$ and thus the flow over the hump will be at critical depth $y_c$. Right after the hump the flow will have the alternate depth ($y_2^{\prime }$) associated with the energy of the upstream flow ($E_2$) and the flow will be supercritical.

Transition of the flow from a supercritical to subcritical regime will occur through a sharp jump in the flow, called hydraulic jump, over which a significant amount of flow energy will be dissipated.

where $y_2$ is the depth of the subcritical flow right before the hump and $y_2^{\prime }$ is the depth of the supercritical flow right after the hump.

## 2-3- Conservation of Momentum in Open Channel Flows:

<div style="text-align: center;">
  <img src="Figures/Ch8/Ch8F13.png" alt="Figure 13" style="width: auto; height: 300px;">
</div>

<p style="text-align: center;"> Schematic showing conservation of momentum in an open channel flow </p>

Before delving into the conservation of momentum, let us find the momentum correction factor for a flow with non-uniform cross sectional velocity distribution.

$$(pvdA)v=\rho v^2 dA$$

$$\beta \rho V^2 A=\int_A \rho v^2 dA$$

$$\beta =\frac{1}{AV^2 }\int_A v^2 dA$$

where $V$ is the mean velocity. 

Thus, the conservation of momentum can be written as follows:

$$\sum F_x =\sum_i \beta_i (\rho V_i .A)*V_i =\rho Q(\beta_2 V_2 -\beta_1 V_1 )$$

Note that $V_i \ldotp A=|V_1 ||A|\cos \theta$ is a dot product, where $\theta$ is the angle between the velocity vector and the surface normal vector. Thus $\theta =180$ for the inflow $\left(-\rho Q\beta_1 V_1 \right)$ and $\theta =0$ for the outflow $\left(\rho Q\beta_2 V_2 \right)$ of the control volume.

Based on the schematic in the last Figure, we have

$$\sum F=F_1 -F_2 -F_f -F_h -F_{air} +Wsin\theta$$

which can be rearranged as follows:

$$F_1 -F_2 -F_{ext} =\rho Q(\beta_2 V_2 -\beta_1 V_1 )$$

where $F_{\textrm{ext}} =F_f +F_h +F_{\textrm{air}} -\textrm{Wsin}\theta$

We know from the law of hydrostatic pressure that $F_1 =\gamma A_1 \bar{y_1 }$ and $F_2 =\gamma A_2 \bar{y_2 }$, where

- $A:$ flow cross section
- $y:$ depth of centroid
- $\gamma$: specific weight of the fluid

Therefore,

$$F_{ext} =(\rho Q\beta_2 V_2 +\gamma \overline{y_2 } A_2 )-(\rho Q\beta_1 V_1 +\gamma \overline{y_1 } A_1 )$$

Assuming $\beta_1 =\beta_2 =1$, we get

$$\frac{F_{ext} }{\gamma }=F=\left(\frac{Q^2 }{gA_2 }+\overline{y_2 } A_2 \right)-\left(\frac{Q^2 }{gA_1 }+\overline{y_1 } A_1 \right)$$

When the value of $F_{\textrm{ext}}$ is much smaller than the right hand side of the above equation, we have:

$$\frac{Q^2 }{gA_2 }+\overline{y_2 } A_2 =\frac{Q^2 }{gA_1 }+\overline{y_1 } A_1$$

The following quantity is called the specific force,

$$F=\frac{Q^2 }{gA}+\overline{y} A$$

Setting $\frac{\partial }{\partial y}F=0$, one can obtain

$$\frac{\partial F}{\partial y}=-\frac{Q^2 }{g}\left(\frac{\frac{dA}{dy}}{A^2 }+\frac{d(\overline{y} A)}{dy}\right)=0$$

Note: $d\left(\bar{y} A\right)\approx \textrm{Ady}$

$$\frac{\partial F}{\partial y}=-\frac{Q^2 }{g}\left(\frac{T}{A^2 }+A\right)=0$$

which results in the critical flow condition as,

$$\frac{Q^2 }{g}=\frac{A^3 }{T}$$


Using $Q=\textrm{VA}$ and $D=\frac{A}{T}$, the above equation reduces to $V=\sqrt{\textrm{gD}}$. Therefore, in a uniform flow with a constant discharge, the specific force is minimum at the critical flow condition, when $\textrm{Fr}=1$.

We already discussed that when a supercritical flow meets a subcritical flow due to continuity of mass and momentum, the flow needs to adjust its shape and a turbulent hydraulic jump occurs during the transition from a super to subcritical flow condition.

Since hydraulic jump is often a local phenomenon that occurs over a short reach of the river flow, we can assume that there is no significant external force acting on the flow in the middle of the jump. In this case, we can assume the specific force before and after the jump remains constant. It should be noted that due to dissipation of turbulence (internal friction) a large amount of the specific energy of the flow is lost and $E_2<E_1$.

As the specific force remains constant before and after the jump, in a uniform rectangular channel flow the above **conjugate depths** (depth before and after a hydraulic jump) can be obtained as follows:

$$y_2 =\frac{-y_1 }{2}+\sqrt{{\left(\frac{y_1 }{2}\right)}^2 +\frac{2q^2 }{gy_1 }}$$
$$y_1 =\frac{-y_2 }{2}+\sqrt{{\left(\frac{y_2 }{2}\right)}^2 +\frac{2q^2 }{gy_2 }}$$

where $q$ is the flow discharge per unit width of the channel. 

During the hydraulic jump, a significant amount of the flow energy can be lost due to dissipative effects of turbulent eddies. This energy loss can be obtained from the specific energy equation $\Delta E=E_2 -E_1$ as follows:

$$\Delta E=\frac{(y_2 -y_1 )^3 }{4y_1 y_2 }$$

**Proof of the conjugate depth:** 

Let us begin by expanding the specific force equation for a rectangular channel

$$F=\frac{Q^2 }{gA}+\overline{y} A$$

For a rectangular channel

$$F=\frac{q^2 b^2 }{gby}+\frac{y}{2}*by=\frac{q^2 b}{gy}+b\frac{y^2 }{2}$$

Since the specific force does not change within a hydraulic jump $F_1 =F_2$, we have

$$\frac{q^2 }{gy_1 }+\frac{y_1^2 }{2}=\frac{q^2 }{gy_2 }+\frac{y_2^2 }{2}$$

Rearranging the above equilibrium, we get,

$$\frac{q^2 }{g}\left(\frac{1}{y_1 }-\frac{1}{y_2 }\right)=\frac{1}{2}(y_2^2 -y_1^2 )\Rightarrow \frac{q^2 }{g}=\frac{1}{2}y_1 y_2 (y_2 +y_1 )$$

The above equation can be expanded as follows at two sections:

$$\frac{V_1^2 y_1^2 }{g}=\frac{1}{2}y_1 y_2 (y_2 +y_1)\;\;\;\;\;\;\; \frac{V_2^2 y_2^2 }{g}=\frac{1}{2}y_1 y_2 (y_2 +y_1 )$$

Thus, one can get

$$\frac{V_1^2 }{gy_1 }=\frac{1}{2}\left(\frac{y_2 }{y_1 }\right)\left(\frac{y_2 }{y_1 }+1\right)=F_{r1}^2 \;\;\;\;\;\;\; \frac{V_2^2 }{gy_2 }=\frac{1}{2}\left(\frac{y_1 }{y_2 }\right)\left(\frac{y_1 }{y_2 }+1\right)=F_{r2}^2$$

Positive roots of the above quadratic equations gives rise to the conjugate depths

$$\frac{y_2 }{y_1 }=\frac{1}{2}\left(\sqrt{1+8Fr_1^2 }-1\right)\Rightarrow y_2 =\frac{-y_1 }{2}+\sqrt{{\left(\frac{y_1 }{2}\right)}^2 +\frac{2q^2 }{gy_1 }}$$

$$\frac{y_1 }{y_2 }=\frac{1}{2}\left(\sqrt{1+8Fr_2^2 }-1\right)\Rightarrow y_1 =\frac{-y_2 }{2}+\sqrt{{\left(\frac{y_2 }{2}\right)}^2 +\frac{2q^2 }{gy_2 }}$$

where ${\textrm{Fr}}_1 =\frac{V_1 }{\sqrt{{\textrm{gy}}_1 }}$ and ${\textrm{Fr}}_2 =\frac{V_2 }{\sqrt{{\textrm{gy}}_2 }}$.
