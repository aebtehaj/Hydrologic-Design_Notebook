# 1- Introduction

**Evapotranspiration (ET)** is the processes by which liquid water at the Earth's surface is transformed into water vapor. Direct transformation from liquid to vapor over a water surface or bare soil is called **evaporation**, while vaporization of water due to a plant's metabolism and growth is called **transpiration**.

ET, though a relatively slow flux, is one of the most important elements of the hydrologic water cycle. ET controls water mass and energy transfer within **land-vegetation-atmosphere** continuum. Additionally, ET contributes significantly to **freshwater losses** in water resources and agricultural systems. Therefore, effective management of ET can significantly improve water and food security.

ET is primarily a water vapor mass flux and thus its measurement is not trivial, especially at the large scale. ET requires the following ingredients to occur:

- **An energy source** (e.g., Sun's radiative energy, wind kinetic energy)
- **Liquid/solid water** (e.g., lakes, soil moisture, snow-covered surfaces, etc.)
- **A transport mechanism** (e.g., molecular diffusion, convection, conduction)

# 2- Heat Transfer in Hydrology

## 2-1 Sensible heat

It is now time to expand on the surface energy balance equation, we covered at the end of Chapter 3.  As you recall, the ($R_n$) at the surface can be partitioned as:

$$R_n =LE+H+G$$     

where $\textrm{LE}\;\left(\lambda \;E\right)$ is the **latent heat flux**, $H$ is the **sensible heat flux**, and $G$ is the **ground heat flux**.

The **heat** is the part of a substance's (water) internal energy that is proportional to its temperature, hence we can sense it.  Sensible heat exchange between a substance (system) and its surroundings (environment) changes its temperature. In terms of the land surface energy budget, sensible heat is the energy transfer largely via convection and turbulence, while the ground heat flux  is largely due to conduction.

  <img src="Figures/Ch4/Ch4F4.png" alt="Figure 4" style="width: auto; height: 300px;">
</div>

The total heat exchange for a system due to a temperature change is defined as:

$$dQ=c_p \,m\,dT$$ 

$m$: mass [kg],  $\textrm{dT}$: temperature change [K],  $c_p$: specific heat $\left\lbrack \mathrm{J}\;{\mathrm{kg}}^{-1} {\mathrm{K}}^{-1} \right\rbrack$. 

In practice, for hydrologic applications, we use the specific heat at constant pressure since the atmospheric pressure changes little over small distances.


## 2-2 Latent Heat

The energy needed to change the phase of a substance (solid, liquid, vapor) without any change in its temperature is called the latent heat of phase change.  Here, are typical values we need for hydrologic applications:

- **Latent Heat of Fusion:** Melting from solid to liquid ${\mathcal{L}}_{s\ell } =3.34\times 10^5 \;{[\mathrm{J}\,\mathrm{k}{\mathrm{g}}^{-1} ]}$ 
- **Latent Heat of Vaporization:** Change from \textit{liquid to water vapor} ${\mathcal{L}}_{\ell v} =2.50\times 10^6 \;{[\mathrm{J}\,\mathrm{k}{\mathrm{g}}^{-1} ]}$ @ 15 $[^{\circ } C]$ 
- **Latent Heat of Sublimation:** Change from ice to vapor ${\mathcal{L}}_{iv} =2.85\times 10^6 \;[{\mathrm{J}\,\mathrm{k}{\mathrm{g}}^{-1} ]}$ 


Primarily, we are interested in ${\mathcal{L}}_{\ell v}$ for ET, which is dependent on temperature as described below:

$${\mathcal{L}}_{\ell v} =2.5\times 10^6 -2370T\;{[\mathrm{J}\mathrm{k}{\mathrm{g}}^{-1} ]}$$

where $T\;[^{\circ } C]$ is temperature in Celsius.  For most practical applications in hydrology $2\ldotp 5\times {10}^6 \;\left\lbrack \mathrm{J}\;{\textrm{kg}}^{-1} \right\rbrack$ serves the purpose. The **sensible and latent heat transfer** mechanisms are the two most important components of heat energy exchange in hydrologic systems. 

## 2-3 Heat Transfer Mechanisms
**Note:** The **flux** of a quantity is the transfer of that quantity per unit area per unit time. 

- **Diffusion or conduction:** Diffusion of water mass or conduction of heat or momentum are due to random molecular motion.  The transfer occurs from areas of higher concentration of the variable of interest to areas of lower concentration.  Heat, momentum and mass molecular transport mechanisms are typically explained via conduction, viscosity, and diffusion coefficients, respectively.  
- **Advection:** Transfer of heat, momentum or mass due to the **bulk motion of fluid parcels**. 
- **Turbulent diffusion:** Turbulence is irregular swirls of fluid motion. In atmosphere, it is mainly generated by mechanical shear due to wind velocity or thermally driven buoyant forces, leading to heat, mass, and momentum transfer. Turbulent diffusion is similar to molecular diffusion; however, in turbulent diffusion we are dealing with random motion of the fluid parcels beyond the effects of molecular diffusion. 
- **Convection:** Combination of advection and turbulent diffusion in transfer of heat, mass and momentum. The amount of heat, moisture, and momentum transport in the atmosphere are measured by their fluxes. 

### 2-3-1 Advective Fluxes: 

As mentioned, advection is the transport of heat, mass, and momentum by bulk fluid motion due to its velocity. 

- **Mass Flux:** $\rho u$  $\left\lbrack \textrm{kg}\;{\mathrm{m}}^{-2} {\mathrm{s}}^{-1} \right\rbrack$ 
- **Momentum Flux:** $\rho u^2$ $\left\lbrack {\textrm{Nm}}^{-2} \right\rbrack$ 
- **Heat Flux:** $\rho_a \,c_p \,T\,u\;$ $\left\lbrack \mathrm{W}\;{\mathrm{m}}^{-2} \right\rbrack$ 
- **Moisture Flux:** $\rho_{a\;} q_{v\;} u$ $\left\lbrack \textrm{kg}\;{\mathrm{m}}^{-2} {\mathrm{s}}^{-1} \right\rbrack$ 
- **Moisture Energy Flux:** ${\mathcal{L}}_{\ell v} \,\rho_a \,q_v \,u$  $\left\lbrack \mathrm{W}\;{\mathrm{m}}^{-2} \right\rbrack$ 

where $\rho_a$ $\left\lbrack \textrm{kg}\;{\mathrm{m}}^{-3} \right\rbrack$ is air density, $u$ $\left\lbrack {\textrm{ms}}^{-1} \right\rbrack$ is the average streamwise velocity, $T_a$ [K] is the air temperature, $q_v$ $\left\lbrack \frac{{\textrm{kg}}_{\textrm{water}\;} }{{\textrm{kg}}_{\textrm{air}} }\right\rbrack$ is the specific humidity and ${\mathcal{L}}_{\ell v}$ $\left\lbrack \mathrm{J}{\textrm{kg}}^{-1} \right\rbrack$ is the latent heat of vaporization. 

### 2-3-2 Diffusive and Turbulent Fluxes: 

For ET, we are primarily interested in **convective transport** of heat, mass and momentum through **turbulent diffusion**. A first-order approximation of the turbulent fluxes can be expressed similar to the way we quantify molecular diffusion in the sense that the fluxes are proportional to the **gradient of the quantity** of interest. 

**a) Momentum Flux:**

Newton's Law of Viscosity:  $\tau =-\mu \frac{du}{dz}=-\rho_a \,\nu \frac{du}{dz}$    $\left\lbrack \mathrm{N}\;{\mathrm{m}}^{-2} \right\rbrack$

- $\tau$: momentum flux (shear stress) $\left\lbrack \mathrm{N}\;{\mathrm{m}}^{-2} \right\rbrack$ 
- $\mu$: dynamic viscosity $\left\lbrack \mathrm{N}\;\mathrm{s}\;{\mathrm{m}}^{-2} \right\rbrack$ 
- $u$: velocity  $\left\lbrack {\mathrm{m}\;\mathrm{s}}^{-1} \right\rbrack$ 
- $\rho_a$: air density $\left\lbrack \textrm{kg}\;{\mathrm{m}}^{-3} \right\rbrack$  
- $\nu$: kinematic viscosity $\left\lbrack {\mathrm{m}}^2 {\mathrm{s}}^{-1} \right\rbrack$ 
- $z$: distance [m]  
- $\frac{du}{dz}$: vertical velocity gradient $\left\lbrack {\mathrm{s}}^{-1} \right\rbrack$ 

**Turbulent Momentum Flux or Reynold's Stress:** $\tau =-\rho_a K_M \frac{du}{dz}$ $\left\lbrack \mathrm{N}\;{\mathrm{m}}^{-2} \right\rbrack$

- $K_M$: momentum eddy diffusivity $\left\lbrack {\mathrm{m}}^2 {\mathrm{s}}^{-1} \right\rbrack$ where $K_M \gg \nu$

**b) Heat Flux:**

Fourier's Law of Conduction: $H=-D_H \frac{dT}{dz}=-\rho_a c_p \alpha \frac{dT}{dz}$    $\left\lbrack \mathrm{W}\;{\mathrm{m}}^{-2} \right\rbrack$

- $H$: Heat Flux $\left\lbrack \mathrm{W}\;{\mathrm{m}}^{-2} \right\rbrack$ 
- $D_H$: thermal conductivity $\left\lbrack W\;{\mathrm{m}}^{-1} {\mathrm{K}}^{-1} \right\rbrack$ 
- $T$:  temperature $\left\lbrack \mathrm{K}\right\rbrack$ 
- $z$: distance $\left\lbrack \mathrm{m}\right\rbrack$ 
- $c_p$: specific heat $\left\lbrack {\textrm{Jkg}}^{-1} \;{\mathrm{K}}^{-1} \right\rbrack$ 
- $\alpha \;$ : thermal diffusivity $\left\lbrack {\mathrm{m}}^2 {\mathrm{s}}^{-1} \right\rbrack$  
- $\rho_a$: air density $\left\lbrack \textrm{kg}\;{\mathrm{m}}^{-3} \right\rbrack$  


**Turbulent Sensible Heat Flux:** $H=-\rho_a c_p K_H \frac{dT}{dz}$

- $K_H$: heat eddy diffusivity $\left\lbrack {\mathrm{m}}^2 {\mathrm{s}}^{-1} \right\rbrack$ where $K_H \gg \alpha$

**c) Moisture Flux:**

Fick's Law of Molecular Diffusion: $E=-\rho_a D_E \frac{dq_v }{dz}$  $\left\lbrack \textrm{kg}\;{\mathrm{m}}^{-2} \;{\mathrm{s}}^{-1} \right\rbrack$

- $\rho_a$: air density  $\left\lbrack \textrm{kg}\;{\mathrm{m}}^{-3} \right\rbrack$
- $D_E$: molecular diffusivity of water vapor in air $\left\lbrack {\mathrm{m}}^2 {\mathrm{s}}^{-1} \right\rbrack$  
- $q_v$: specific humidity  $\left\lbrack \frac{{\textrm{kg}}_{\textrm{watar}\;} }{{\textrm{kg}}_{\textrm{air}} }\right\rbrack$ 
- $z$: distance $\left\lbrack \mathrm{m}\right\rbrack$ 

**Turbulent Evaporation Flux :** $E=-\rho_a K_E \frac{dq_v }{dz}$

- $K_E$: moisture eddy diffusivity $\left\lbrack {\mathrm{m}}^2 {\mathrm{s}}^{-1} \right\rbrack$ where $K_E \gg D_M$ 

As we have discussed the turbulent diffusivity coefficients are much larger than the **molecular diffusion coefficients**. This is true, especially in the atmospheric boundary layer, and thus we typically neglect the molecular diffusion portion of the transport.

Some experimental evidence suggests that in a neutral atmosphere (no density gradient) $K_H =K_E =1.35K_M$, which will be defined in the next slide. However, often in practice these three eddy diffusivity values maybe assumed to be equal in absence of additional a priori evidence. Note that these fluxes are computed in a layer of air near the earth surafce within the **atmospheric boundary layer**.

# 3- Atmospheric Boudary Layer

Evapotranspiration fluxes are largely due to turbulent heat and moisture fluxes in the **atmospheric boundary layer (ABL)**, which is the lowest part of the atmosphere where the air flow properties are highly affected by friction due to interaction of the air flow with surface roughness.  The depth of the ABL is typically around 1-2 km.  


- **Neutral ABL:** The air density is uniform throughout the ABL depth, meaning if we displace a parcel of air, it will stay in its new position. 
- **Stable ABL:** Denser air is below less-dense air through the entire ABL. If we displace an air parcel, it tends to return to its original position.
- **Unstable ABL:**  Less-dense air is below denser air. If we displace a parcel of air, it may not return to its original position.

**Why is stability of ABL important?** 

When the atmosphere is **stable**, the turbulent moisture and heat fluxes are **suppressed**, whereas in unstable atmospheres, they are enhanced. Therefore, **ET is a function of atmospheric stability** condition.

There are two main turbulent transport mechanisms that drive ABL instability: 

- **Static instability:** Buoyancy driven turbulence (buoyancy $\gg$ wind shear kinetic energy ) 
- **Dynamic instability:** Wind shear driven turbulence (wind shear $\gg$ buoyancy kinetic energy) 

**How can we characterize stability of ABL?**

There are several **dimensionless parameters** that are commonly used to define atmospheric stability from data. The **gradient Richardson number** is a dimensionless parameter encoding the ratio of the buoyancy versus shear production of energy of tubulence:

$$R_i =\frac{\frac{g}{T_v }.\frac{\partial T_v }{\partial z}}{\left\lbrack {\left(\frac{\partial u}{\partial z}\right)}^2 \right\rbrack }= \frac{\rm Buoyant \, turbulence \,production}{\rm Shear \, turbulence \, production}$$

where $z$ denotes vertical direction, $T_v$ is the virtual temperature, $g$ is gravitational acceleration, and $u$ is the horizontal wind velocity. The values of $R_i \;$ are interpreted as 

- **Unstable:** $R_i <0\;$ 
- **Neutral Condition:** $R_i =0$ 
- **Stable:** $R_i >0$ 

Another important stability parameter in the ABL is the so-called **Obukhov Length** $L$: 


$$L=\frac{-T_v \,u_*^3 }{\kappa \,g\,\left(\frac{H}{c_p \,\rho_a }\right)}~~~~\textrm{[m]}$$

where $T_v$ is the mean virtual temperature near the earth surafce, $u_* =\sqrt{\frac{\tau }{\rho_a }}$ is the shear velocity, where $\tau$ is the shear stress $\left\lbrack \mathrm{N}\;{\mathrm{m}}^{-2} \right\rbrack$ and $\rho_a$ denotes the air density, $\kappa \simeq 0.41$ is the *Von Karman* constant, $g$ is the earth gravitational acceleration and $H$ is the sensible heat flux $\left\lbrack \mathrm{W}\;{\mathrm{m}}^{-2} \right\rbrack$ at the surface.

The Obukhov length ($L$) is an important \textbf{scaling variable} used to account for the effects of atmospheric stability condition on momentum, heat and mass fluxes. Typically, we divide the measurement height of the weather station above the surface by $L\;$ to define a dimensionless representation of the Obukhov length:

$$\zeta =\frac{z}{L}$$

where $\zeta \;$ is often called the **surface layer height scaling variable**.

When the Obukhov length is **negative** ($L<0$) the heat flux is upward ($H^{\uparrow } >0$) and thus the **atmosphere is unstableand** when it is **positive** $L>0$, the heat flux is downward ($H^{\downarrow } <0$) and thus the atmosphere is **stable**. 

The gradient Richardson number $R_i$ and $\zeta$ are related to each other through the following semi-empirical relationships:

$$\begin{array}{l}
\zeta =\frac{R_i }{1-5R_i }\,\,~~~~\textrm{stable}\,(\zeta >0;\,\,\,0\le R_i <0.2)\\
\zeta =R_i \,\,\,\,\,\,\,\,\,\,~~~~~~\textrm{unstable}\,(\zeta <0;\,\,\,R_i <0)
\end{array}$$

Before we move forward, it is important to clarify that the above discussion is confined to the **Atmospheric Surface Layer** (ASL).  This layer is typically taken as the bottom 10\% of the ABL where the vertical velocity, temperature, and moisture vary rapidly, while vertical fluxes of heat, moisture and momentum are approximately constant.







## 3-1 Wind Profile in Atmospheric Boudary Layer
### 3-1-1 Log Law in Neutral ABL

In order to understand ET fluxes from the Earth's surface to the ASL, we must have an understanding of the **mean wind velocity profile** as it is the main driver of the turbulent diffusive fluxes. It has been (theoretically) shown that the mean velocity profile in a **neutral** ASL follows a logarithmic shape as follows:


$$\bar{u} =\frac{u_* }{\kappa }\ln \left(\frac{z}{z_0 }\right).$$

where $z_0$ [m] is called the **momentum roughness length** at which the wind velocity is zero, $u_*$ $\left\lbrack {\textrm{ms}}^{-1} \right\rbrack$ is the **friction velocity**, and $\kappa \simeq 0.41$ is the Von Karman constant. 

Over a canopy, where the roughness elements are densely packed, the wind velocity goes to zero at height $d+z_0$, where $d$ is called the **zero-plane displacement height**. As a result, we have

$$\bar{u} =\frac{u_* }{\kappa }\ln \left(\frac{z-d}{z_0 }\right).$$

  <img src="Figures/Ch4/Ch4F7b.png" alt="Figure 7b" style="width: auto; height: 250px;">
</div>

**Note:** that $\bar{\;u}$ denoted the mean values. In reality, if we measure the velocity profile in an instant of time, due to tubulence, the profile will be wiggly.  

As you can see, we have introduced three parameters to describe the velocity profile:

- **Momentum roughness length** ($z_0$):  This parameter describes the roughness of the surface and can be thought of as height above the surface at which the air velocity is zero.  Typical values are shown in the table below and a general approximation is that $z_0 =0.1h$, where $h$ is the average roughness height of surface soil or canopy.
- **Zero-plane displacement** ($d$):  This is an offset to apply the log-law over a dense canopy.  Studies have showen that a good approximation is $d=0.7h$. 
- **Friction velocity** ($u_*$):  Also called **shear velocity**, is an important **velocity scale** that is related to the surface shear stress $\tau_0$  as $u_* =\sqrt{\frac{\tau_0 }{\rho_a }}$. A general rule is that the friction velocity is between 5\% to 10\% of the mean flow velocity.

### 3-1-2 Logarithmic Law in Non-Neutral ASL:

Observations show that in the stable and unstable ASL, the velocity profile deviates from the log-law as shown below.  Now if we differentiate the original log-law, we get:

$$\bar{u} =\frac{u_* }{\kappa }\ln \left(\frac{z-d}{z_0 }\right)~~~~\Longrightarrow ~~~~\frac{d\bar{u} }{dz}=\frac{u_* }{\kappa (z-d)}.$$

To account for the **atmospheric stability** or **instability** conditions, we can multiply the velocity profile by a correction factor $\phi_M \left(R_i \right)$, that can be explained as a function of the **gradient Richardson number**. 

Therefore, the corrected velocity profile can be explained as follows:

$$\frac{d\bar{u} }{dz}=\frac{u_* }{\kappa (z-d)}\phi_M \left(R_i \right).$$

Atmospheric stability also affects the profile of moisture and temperature. There are several parameterizations that correct for the velocity $\phi_M$, temperature $\phi_H$, and moisture $\phi_E$ profiles under non-neutral atmosphere. Here, we focus on the formulation by Dyer and Hicks (1970) as follows:

$$\phi_M^2 =\phi_H =\phi_E =(1-16R_i )^{-0.5} =(1-16\zeta )^{-0.5} ~~~~R_i <0~~\textrm{or}~~\zeta <0~~{(\mathrm{u}\mathrm{n}\mathrm{s}\mathrm{t}\mathrm{a}\mathrm{b}\mathrm{l}\mathrm{e})}$$
$$\phi_M =\phi_H =\phi_E =\frac{1}{1-5R_i }=1+5\zeta ~~~~0\le R_i \le 0.2~~\textrm{or}~~\zeta >0{~~(\mathrm{s}\mathrm{t}\mathrm{a}\mathrm{b}\mathrm{l}\mathrm{e})}$$

Since we have $\tau_0 =\rho_a K_M d\bar{u} /dz$ from the \textit{Newton's Law of Viscosity}, $u_*^2 =\tau_0 /\rho_a$ and $\frac{d\bar{u} }{dz}=\frac{u_* }{\kappa (z-d)}\phi_M$, we can conclude that the momentum diffusivity coefficient is 

$$K_M =\kappa u_* (z-d)\phi_M^{-1} .$$

The same derivation holds for the heat and moisture turbulent diffusivity vlaues for non-neutral atmosphere as follows:

$$K_H =\kappa u_* (z-d)\phi_H^{-1}$$

$$K_E =\kappa u_* (z-d)\phi_E^{-1}$$

which can be used for computation of the turbulent heat and moisture fluxes under stable or unstable atmosphere. Obviously, the correction factors are equal to one for a neutral atmosphere.


## 3-2 Surface Evaporation and Heat Fluxes
### 3-2-1 Neutral Atmopshere

As we have discussed earlier, the moisture and heat flux can be defined by the following turbulent diffusion equations:

$$E=-\rho_a K_E \frac{dq_v }{dz}~~~~{[\mathrm{K}\mathrm{g}\,{\mathrm{m}}^{-2} \,{\mathrm{s}}^{-1} ]}$$
$$H=-\rho_a c_p K_H \frac{dT}{dz}~~~~{[\mathrm{W}{\mathrm{m}}^{-2} ]}$$

where $K_E$ and $K_H \;$ are the respective moisture and heat turbulent diffusivity values $[{{\mathrm{m}}^2 {\mathrm{s}}^{-1} }]$ since the ASL is generally turbulent. 

Now we would like to use these equations in combination with the information gained from the velocity profile to derive an equation for the fluxes in terms of variables we can easily measure.

Under neutral atmosphere with zero displacement ($d=0$), we assume $K_H \simeq K_E \simeq K_M$ and $K_M \simeq \kappa zu_*$. As results one can have

$$E=-\rho_a (\kappa zu_* )\frac{dq_v }{dz}$$

and thus

$$dq_v =-\frac{E}{\rho_a \kappa u_* }\frac{dz}{z}~~\Rightarrow ~~\int_{q_{v0} }^{q_v } dq_v =\frac{-E}{\rho_a \kappa u_* }\int_{z_{0v} }^z \frac{dz}{z},$$

where, $q_{v0}$ is the moisture at height $z_{0v} \;$ near the earth surface. Near the earth surface, when the soil is completley saturated, there is a thin air that fully saturated too. Therefore, over moist soil, when there is no limitation for evaporation of water, $q_{\textrm{v0}}$ is the **saturated specific humidity**. In the above equations, obviously, $q_v$ is the specific humidity at height $z$ above the surface. This height is often 2 or 10 meters above the surface depending on the height of the instrument on a weather station. 

After integration, one can obtain

$$q_v (z)-q_{v0} =-\frac{E}{\rho_a \kappa u_* }\ln \left(\frac{z}{z_{0v} }\right),$$


which can be rearranged as follows with substitution for $u_* =\kappa u(z)/\ln (z/z_{0m} )$ from the the log-law:

$$E=\frac{\rho_a \kappa^2 u(z)}{\ln \left(\frac{z}{z_{v0} }\right)\ln \left(\frac{z}{z_{0m} }\right)}\left(q_{v0} -q_v (z)\right),$$

- $z_{0m}:$ momentum surface roughness length (wind profile goes to zero)
- $z_{0v}:$ moisture surface roughness length (moisture profile goes to near surface $q_{v0}$)
- $q_{0v}:$ near surface specific humidity of air 
- $q_v (z):$ air specific humidity at height $z$ (typically $z=2\;$ or 10 m) 


**NOTE that the bar notation refering to the average wind velocity is dropped for simiplicity. Velocity and humidity values are the average values over time.**

----


**Summary:**

The above equation can have a simpler representation if we assume ${z_0 =z}_{0m} \approx z_{0v}$ as follows:

$$E=\frac{\rho_a \kappa^2 u(z)}{{\left\lbrack \ln \left(\frac{z}{z_0 }\right)\right\rbrack }^2 }\left(q_{v0} -q_v (z)\right)~~{[\mathrm{K}\mathrm{g}{\mathrm{m}}^{-2} {\mathrm{s}}^{-1} ]}$$
$$LE={\mathcal{L}}_{lv} \times E~~[{\mathrm{W}{\mathrm{m}}^{-2} ]}$$

The above moisture flux is often expressed in a more compact and intuitive form as follows, where the displacement height is nonzero.

$$E=\rho_a \,\frac{q_{v0} -q_v (z)}{r_{av} }$$
$$r_{av} =\frac{{\left\lbrack \ln \left(\frac{z-d}{z_0 }\right)\right\rbrack }^2 }{\kappa^2 \,u(z)}$$

where $r_{av} \;$ is called **aerodynamic vapor resistance** [$\mathrm{s}\;{\mathrm{m}}^{-1}$]. 

Similarly, we can also derive an equation for the **sensible heat flux** in neutral condition as follows:

$$H=\rho_a \,c_p \frac{T_0 -T(z)}{r_{ah} }~~{[\mathrm{W}{\mathrm{m}}^{-2} ]}$$
$$r_{ah} =\frac{\ln \left(\frac{z-d}{z_{0m} }\right)\ln \left(\frac{z-d}{z_{0h} }\right)}{\kappa^2 \,u(z)}\approx \frac{{\left\lbrack \ln \left(\frac{z-d}{z_0 }\right)\right\rbrack }^2 }{\kappa^2 \,u(z)}$$

where $r_{ah}$ is called \textbf{aerodynamic heat resistance} $\left\lbrack \mathrm{s}\;{\mathrm{m}}^{-1} \right\rbrack$. 

Estimating the values of $z_{0v} \;\textrm{and}\;z_{0h}$  requires profile data for moisture and heat. These roughness parameters tend to be more variable than $z_{0m}$.  Experimental results suggest that $z_{0v} =z_{0h}$ is agood approaximation and if there is no available heat or moisture profile data, then we may assume $z_{0v} =z_{0h} =0.1z_{0m}$ for research purposes (Allen et al. 2005).  For most hydrologic practical problems, we can assume $z_0 =z_{0m} =z_{0h} =z_{0v}$ in the absence of any additional information. 

---





### 3-2-2 Non-neutral Atmopshere
In a non-neutral atmosphere we just need to update the resistance parameters as follows:

$$H=\rho_a \,c_p \frac{T_0 -T(z)}{r_{ah}^{\prime } }$$
$$E=\rho_a \,\frac{q_{v0} -q_v (z)}{r_{av}^{\prime } }$$

The non-neutral resistance paremeters can be computed as follows (Thom 1975):

$$r_{ah}^{\prime } =r_{ah} \times (\phi_H \phi_M )$$

$$r_{av}^{\prime } =r_{av} \times (\phi_E \phi_M ).$$

The above formulas for evaporative fluxe assume that there are enough water and soils and there is not any water limitations. In other words, evaporation is in its potential condition -- called **potential or reference evaporation**. However, in reality soil is not always saturated and there are limitations in terms of the water vailability that can supress the potential evaporation. 

### 3-2-3 Soil Moisture Effects on Evaporation 

For a saturated soil surface or open water, the evaporation is at its potential rate $E_p$ and it is reasonable to assume that $q_{v0} =q_{vs}$, where $q_{vs}$ is saturated specific humidity. For unsaturated soils, the soil moisture water content is denoted by $\theta =\frac{V_w }{V_t }$  ${[\mathrm{c}{\mathrm{m}}^3 \mathrm{c}{\mathrm{m}}^{-3} ]}$ where $V_w$ is the water volume and $V_t$ is the total volume of the soil metrics. In this case the actual evaporation can be quantified as follows: 

$$E=\beta (\theta )E_p$$

where $\theta$ soil moisture content ${[\mathrm{c}{\mathrm{m}}^3 \mathrm{c}{\mathrm{m}}^{-3} ]}$ and $0\le \beta (\theta )\le 1$ is a  correction factor, which is a function of the soil type and moisture content. 


$$\beta \left(\theta \right)=\left\lbrace \begin{array}{cc}
1 & \,\,\theta \ge \theta_{fc} \\
\frac{\theta -\theta_{wp} }{\theta_{fc} -\theta_{wp} } & \,\,\theta_{wp} \le \theta \le \theta_{fc} \\
0 & \,\,\theta \le \theta_{wp} .
\end{array}\right.$$

where $\theta_{\textrm{fc}}$ is the soil moisture at **field capacity** and $\theta_{\textrm{wp}}$ is the soil moisture at **wilting point.** We will learn how to comute the soil moisture for these two conditions in the next chapter.

## 3-3 Evapotranspiration Processes

Plant's synthesize visible solar energy (photosynthesis) for their metabolism. Throughout this process, water is evaporated by areal parts of the plant such as leaves, stems, etc. The process of water vaporization from soil to atmosphere through plant's metabolism is called **transpiration**. On plant's leaves there are pores called **stomata** that allow plants to uptake CO2 for photosynthesis and release oxygen and water vapor to the atmosphere through the process of transpiration, which reduces the plants temperature. The process and rate of transpiration are controlled by the dynamics of stomata. 

<div style="text-align: center;">
  <img src="Figures/Ch4/Ch4F11.png" alt="Figure 11 style="width: auto; height: 350px;">
</div>

<p style="text-align: center;">Pathway for water loss from surface of a leaf (Credit: Jones, 1983).
</p>


Stomata respond to environmental forcings (e.g., air temperature, humidity, wind speed, sunlight intensity, soil water supply, etc.) by **changing their opening size**. When plants are under stress, typically stomata close their aperture to protect the plants against excessive water loss and wilting. As a general rule, larger leaves have more stomota, which increases the ET rate from the leaves. Some plants have a waxy cuticle that reduces ET from the leaf surface (e.g., Xerophyte plants) such as the Cactus family. As noted, environmental stress affects plant's transpiration rate as follows:


- Increase of light (net radiation) increases the transpiration up to an asymptotic limit. 
- Transpiration increases and reaches to a maximum for an optimal temperature and then reduces as tempertaure increases.
- Increase of water vapor pressure deficit often decreases the ET. In other words, when the air becomes so drt, plants tend to keep the water and traspire less. 
- Increase of soil moisture increases the rate of transpiration.

In general, plants naturally attempt to increase their survivability in response to environmental stresses. The guard cells that control the size of the stomatal openings play a very critical and complex regulatory role. For instance when temperature increases above the tolerance limits of the plant, the stomatal openings start to reduce their sizes to reduce the rate of evaporation and increase the chance of survival. This complex regulatory role is often simplified and parameterized  through the **canopy resistance factor**.

The **canopy resistance** $r_c$ can be modeled based on the stomatal resistance as follows:

$$r_c =\frac{\;r_s }{\textrm{LAI}}\;\;\;\left\lbrack {\mathrm{s}}^{-1} \;\mathrm{m}\right\rbrack$$

where $r_s$ is the **stomatal resistance** at a leaf level and $r_c$ is the **canopy resistance**, and LAI is the so-called Leaf Area Index. 

The LAI is defined as the \textit{one-sided green leaf area per unit ground surface area }and ranges from $0<\textrm{LAI}<10$, where the upper bound refers to dense conifer forests.

<div style="text-align: center;">
  <img src="Figures/Ch4/Ch4F12.png" alt="Figure 12 style="width: auto; height: 350px;">
</div>

As mentioned, there are many forcings that impact stomata behavior at leaf level and thus $r_s$.  One can model stomata resistance at a leaf-scale as follows:

$$r_s =\frac{r_{min} }{f_{R_s } \cdot f_{T_a } \cdot f_{\delta_e } \cdot f_{\theta } }$$

where $r_{min}$ is the minimal stomatal resistance determined from experiments on different plant types. The correction parameters ($f$) may be parameterized as follows:


- **Radiation:** $f_{R_s } =\frac{1.105R_{is} }{1.007R_{is} +104.4}$, where  $R_{is}$ is incident shortwave radiation ${[\mathrm{W}{\mathrm{m}}^2 ]}$ .  
- **Air water vapor deficit:** $f_{\delta_e } =1-0.00023\delta_e \textrm{,}$ where $\delta_e =e_s (T_a )-e(T_a )$  ${[\mathrm{N}\,{\mathrm{m}}^{-2} ]}$. 
- **Temperature:** $f_{T_a } =\frac{T_a (40-T_a )^{1.18} }{690}$, where $0\le T_a \le 40$ ${[^{\circ } \mathrm{C}]}$
- **Soil moisture:** $f_{\theta } =\left\lbrace \begin{array}{cc}
1 & \,\,\theta \ge \theta_{fc} \\
\frac{\theta -\theta_{wp} }{\theta_{fc} -\theta_{wp} } & \,\,\theta_{wp} \le \theta \le \theta_{fc} ~~\\
0 & \,\,\theta \le \theta_{wp} 
\end{array}\right.$ 




```python
#Import Modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['font.size'] = 20

R_is = np.arange(0, 1001, 50)  # [W/m2]
delta_s = np.arange(0, 5001, 50)  # [Pa]
T_a = np.arange(0, 41, 1)  # [C]
theta_fc = 0.4
theta_wp = 0.1

# Calculate f_Rs
f_Rs = 1.105 * R_is / (1.007 * R_is + 104.4)

# Calculate f_delta_s
f_delta_s = 1 - 0.00023 * delta_s
f_delta_s[f_delta_s < 0] = 0

# Calculate f_Ta
f_Ta = (T_a * (40 - T_a) ** 1.18) / 690

# Calculate f_theta
theta_values = np.arange(0, 0.51, 0.01)
f_theta = np.zeros_like(theta_values)
for i, theta in enumerate(theta_values):
    if theta <= theta_wp:
        f_theta[i] = 0
    elif theta >= theta_fc:
        f_theta[i] = 1
    else:
        f_theta[i] = (theta - theta_wp) / (theta_fc - theta_wp)

# Plotting
fig, axs = plt.subplots(1, 4, figsize=(15, 5))

axs[0].plot(R_is, f_Rs, linewidth=2, color='blue')
axs[0].grid(True)
axs[0].set_xlabel(r'$R_s$ [W m$^{-2}$]')
axs[0].set_ylabel(r'$f_{R_s}$')

axs[1].plot(delta_s, f_delta_s, linewidth=2, color='red')
axs[1].grid(True)
axs[1].set_xlabel(r'$\delta_s$ [Pa]')
axs[1].set_ylabel(r'$f_{\delta_s}$')
axs[1].set_xlim([0, 5000])

axs[2].plot(T_a, f_Ta, linewidth=2, color='green')
axs[2].grid(True)
axs[2].set_xlabel(r'$T_a$ [°C]')
axs[2].set_ylabel(r'$f_{T_a}$')
axs[2].set_ylim([0, 1])

axs[3].plot(theta_values, f_theta, linewidth=2, color='black')
axs[3].grid(True)
axs[3].set_xlabel(r'$\theta$ [-]')
axs[3].set_ylabel(r'$f_{\theta}$')

plt.tight_layout()
plt.show()
```

The overall resistance to evapotranspiration is a combination of both **canopy resistance** $r_c$, that accounts for the various vapor transport mechanisms within the canopy and soil surface and the **aerodynamic vapor resistance** $r_{\textrm{av}}$ that accounts for the effects of the wind velocity and turbulent transport.  Since these resistances are in series, we can sum them up into a single term:

$$r_{ac} =r_c +r_{av} ~~~~{[\mathrm{s}{\mathrm{m}}^{-1} ]}$$

## 3-4 Models of Evapotranspiration

Apart from the pants regulations of water loss from surface, there are two main mechanisms controlling surface evaporation fluxes: (i) supply of energy to provide the latent heat of vaporization and (ii) the ability to transport the vapor away from the evaporative surface. Solar radiation is the mean sources of heat energy. The ability to transport vapor away from the surface largely depends on turbulence kinetic energy and thus wind velocity neat the surface as well as the gradient of moisture near the surface. Higher wind velocity often increases the turbulent kinetic energy and moves saturated air parcel form near the surface. When the overlying air parcels are dry and soil is wet, the higher moisture gradient accelerates and increases the magnitude of evaporative fluxes. 

There are three types of ET models:

- **Energy balance models**
- **Aerodynamic models**
- **Combined energy balance and aerodynamic models**


### 3-4-1 Energy balance models

So far, we have focused primarily on flux-based or aerodynamic methods for calculating sensible and latent heat fluxes.  However, it is also very common to further constrain the calculated ET flux, based on the surface energy balance (SEB).

$$R_n =LE+H+G$$

By convention, net radiation $R_n$, is positive when toward the surface, whereas $H$, $\textrm{LE}$, and $G$ are positive away from the surface. 

<div style="text-align: center;">
  <img src="Figures/Ch4/Ch4F15.png" alt="Figure 15 style="width: 300px; height: auto;">
</div>

<p style="text-align: center;">Schematic of typical diurnal variation of the surface energy balance for a well watered soil surface (Stull, 2015).
</p>

Clearly $\textrm{LE}=R_{n\;} -H-G$. If both sensible and ground heat fluxes are zero, Then the rate of evaporative water loss for a surface with unit area of 1 $\left\lbrack {\mathrm{m}}^2 \right\rbrack$ is as follows:

$$E=\frac{\;\textrm{LE}}{\mathcal{L}{\;}_{\ell v} \rho {\;}_w }=\frac{\;R_n -H-G}{\mathcal{L}{\;}_{\ell v} \rho {\;}_w }\simeq \frac{\;R_n }{\mathcal{L}{\;}_{\ell v} \rho {\;}_w }\left\lbrack \mathrm{m}\;{\mathrm{s}}^{-1} \right\rbrack$$

where $\rho {\;}_w$ is the water density 1000 $\left\lbrack \textrm{kg}\;{\mathrm{m}}^3 \right\rbrack$. Note that $E\;\left(\frac{\;\mathrm{m}}{\mathrm{s}}\right)\times \;1000\left(\frac{\;\textrm{mm}}{\mathrm{m}}\right)\times \;8600\left(\frac{\;\mathrm{s}}{\textrm{day}}\right):=E\;\left(\frac{\;\textrm{mm}}{\textrm{day}}\right)$  

____
**Example Probelm 4.1.:** Calculate by the energy balance model of evaporation rate from an open water surface, if the radiation is 200 [w/m2] and the air temperature is 25 [C], assuming no sensible and ground heat flux.

**Solution:** we have ${\mathcal{L}}_{\ell v\;} =2500-2\ldotp 36\times 25=2441\;\;\left\lbrack \frac{\textrm{KJ}}{\textrm{kg}}\right\rbrack \ldotp \;$ Therefore,

$$E=\frac{\;200}{2441\times {10}^3 \times 1000}=8\ldotp 2\times {10}^{-8} \;\;\;\left\lbrack {\textrm{ms}}^{-1} \right\rbrack$$

Let us covert the value to a more tangible unit:

$$E=8\ldotp 2\times {10}^{-8} \left(\frac{\mathrm{m}}{\mathrm{s}}\right)\times 1000\;\left(\frac{\textrm{mm}}{\mathrm{m}}\right)\times 86400\;\left(\frac{\mathrm{s}}{\textrm{day}\;}\right)=7\ldotp 10\;\;\;\left\lbrack \textrm{mm}\;{\textrm{day}}^{-1} \right\rbrack$$

----

### 3-4-2 Aerodynamic models
Beside the supply of energy, another important factor controlling the evaporation processes is the ability of the overlying atmosphere to transport water away from the surface. As we already explained, the transport rate depends on the humidity gradient and the turbulent kinetic energy quantified through the wind speed. This class of methods that calculate the evaporation fluxes solely based on the wind speed and moisture gradient is often called aerodynamic methods. We already derived the formula as follows for mass flux of water vapor:

 $$E=\rho_a \,\frac{q_{v0} -q_v (z)}{r_{av} }=\rho_a \frac{\epsilon }{P}\frac{(e_{s0} -e_a (z))}{r_{av} }~~{[\mathrm{K}{\mathrm{g}}_{\mathrm{w}\mathrm{a}\mathrm{t}\mathrm{e}\mathrm{r}} {\mathrm{m}}^{-2} {\mathrm{s}}^{-1} ]}$$

$$r_{av} =\frac{{\left\lbrack \ln \left(\frac{z-d}{z_0 }\right)\right\rbrack }^2 }{\kappa^2 \,u(z)}$$

where we considered $q_{v\;} =\epsilon \frac{\;e}{P}$. Note that for computing the the rate in terms of $\left\lbrack \textrm{mm}\;{\textrm{day}}^{-1} \right\rbrack$, we need $E\;\left\lbrack \textrm{mm}\ldotp {\textrm{day}}^{-1} \right\rbrack =:E\;\left\lbrack {\textrm{Kgm}}^{-2} {\mathrm{s}}^{-1} \right\rbrack \times \rho_w^{-1} 1000\;\left(\frac{\textrm{mm}}{\mathrm{m}}\right)\times 86400\;\left(\frac{\mathrm{s}}{\textrm{day}\;}\right)$.

____
**Example Probelm 4.2.:** Calculate the evaporation rate from an open water surface, for a neutral atmosphere, by the aerodynamic method with an air temperature of 25 degrees Celsius, relative humidity of 45\%, air pressure of 101.3 kPa, and wind speed of 2 m/s. All measurements are at height of 2 m above the water surface. Assume roughness height $z_0 =0\ldotp 02$ cm. If throughout the day, the mean gradient Richardson number will be $R_i =0\ldotp 1$, what would be an estimate of the evaporation? Does it increase or decrease? Determine the percentage of change.



**Solution:** 

$$\begin{array}{l}
e_s =611\times \;\exp \left\lbrack \frac{{17\ldotp 27\times \;T}_a }{237\ldotp 3+T_a }\right\rbrack =3168\;\left\lbrack \textrm{Pa}\right\rbrack \\
e_a =\textrm{RH}\times e_s =1426\;\left\lbrack \textrm{Pa}\right\rbrack \\
T_v =\frac{\;T_a }{\left\lbrack 1-\left(1-\epsilon \;\right)\frac{\;e_a }{P}\right\rbrack }=299\ldotp 75\;\left\lbrack K\right\rbrack \\
\rho_a =\frac{\;P}{R_d \times \;T_v }=1\ldotp 1775\;\;\left\lbrack {\textrm{kgm}}^{-3} \right\rbrack \\
r_{\textrm{av}} =\frac{\;\left\lbrack \ln \;{\left(\frac{\;z}{z_0 }\right)}^2 \right\rbrack }{\kappa^2 u\left(z=2\;m\right)}=252\ldotp 3\;\left\lbrack {\textrm{sm}}^{-1} \right\rbrack \;\\
E=\rho {\;}_a \times \frac{\;\epsilon \;\;}{P}\frac{\;\left(e_{\textrm{s0}} -e_a \right)}{r_{\textrm{av}} }\times \rho {\;}_w^{-1} \times \left(\frac{\;1000\;\textrm{mm}}{1\;\mathrm{m}\;}\right)\times \left(\frac{\;86400\;s}{\textrm{day}\;}\right)=4\ldotp 3\;\;\left\lbrack \frac{\textrm{mm}}{\textrm{day}}\right\rbrack 
\end{array}$$


When $0{\le R}_i \le 0\ldotp 2$, the atmosphere is stable and thus we have

$$\phi_M =\phi_H =\frac{\;1}{1-5R_i }=2$$

$$r_{\textrm{av}}^{\prime \;} =r_{\textrm{av}} \times \left(\phi_M \phi_H \right)=1009\;\left\lbrack {\textrm{sm}}^{-1} \right\rbrack$$

$$E_{\textrm{stable}} =E\times \frac{\;r_{\textrm{av}} }{r_{\textrm{av}}^{\prime \;} }=1\ldotp 08\;\left\lbrack \textrm{mm}\;{\textrm{day}}^{-1} \right\rbrack$$

----

```python
# Inputs
rho_w = 1000  # density of water [kg/m^3]
kappa = 0.41  # von Karman constant
RH = 0.45  # relative humidity
z = 2  # measurement elevation [m]
z_0 = 0.02 / 100  # roughness height [m]
eps = 0.622
u_2 = 2  # wind velocity [m/s]
P = 101.3 * 10**3  # Pressure [Pa]
T_a = 25  # air temperature [°C]
R_d = 287  # specific gas constant of dry air [J/kg.K]
R_i = 0.1  # Gradient Richardson number

# Saturated water vapor pressure function
def e_s(T_a):
    return 611 * np.exp(17.27 * T_a / (237.3 + T_a))

# Solution
e_a = RH * e_s(T_a)  # air vapor pressure [Pa]
T_v = (T_a + 273.15) / (1 - (1 - eps) * e_a / P)  # virtual air temperature [K]
rho_a = P / (R_d * T_v)  # air density [kg/m^3]
r_av = (np.log(z / z_0))**2 / (kappa**2 * u_2)  # resistance [s/m]
E = rho_a * (eps / P) * (e_s(T_a) - e_a) / r_av * (1 / rho_w) * 1000 * 86400  # [mm/day]

print(f"Mean daily evaporation, for a neutral atmosphere, by the aerodynamic method is {E:.2f} [mm/day]")
```

```python
# For R_i = 0.1, the atmosphere is stable
phi_M = 1 / (1 - 5 * R_i)
phi_H = phi_M
r_av_prime = r_av * phi_M * phi_H
E_stable = rho_a * (eps / P) * (e_s(T_a) - e_a) / r_av_prime * (1 / rho_w) * 1000 * 86400  # [mm/day]

print(f"Mean daily evaporation, for a stable atmosphere, by the aerodynamic method is {E_stable:.2f} [mm/day]")

# percentage reduction
print(f'percentage reduction is {(E-E_stable)/E*100}')
```

### 3-4-3 Combined models

From SEB, we have one equation and three unknowns ($\textrm{LE},H,G$), considering we can easily estimate or measure $R_n$.  Since the ground heat flux does not vary significantly compared to the sensible and latent heat fluxes, we typically define the available energy that is partitioned to the sensible and latent heat fluxes as follows:

$$Q_n =R_n -G.$$

Experimental evidence suggests that we can assume $G$ is some fraction of $R_n$.  Based on the field data $G\approx 0.05R_n$ over vegetation canopies and $G\approx 0.315R_n$ over the bare soil. This assumption leaves us with one equation and two unknowns that is $Q_n =LE+H$. Therefore, to the solve the land surface energy balance equation, we only need an extra equation. 

**Bowen Ratio (BR) method:** To add an extra equation that enables us to obtain the sensible and latent heat flux, we can use the **Bowen ratio**, which is simply the ratio of sensible to latent heat flux considering $r_{\textrm{ah}} =r_{\textrm{av}}$:

$$\beta =\frac{H}{LE}=\frac{\rho_a c_p \frac{T_0 -T_a }{r_{ah} }}{{\mathcal{L}}_{lv} \rho_a \frac{q_0 -q_{va} }{r_{av} }}=\frac{c_p (T_0 -T_a )}{{\mathcal{L}}_{lv} (q_{v0} -q_{va} )}=\frac{c_p P(T_0 -T_a )}{0.622\times {\mathcal{L}}_{lv} (e_0 -e_a )}$$

where the air tempertaure $T_a$  [C] and vapor pressure $e_a$ [Pa] is often obtain 2 or 10 meters above the surface. A good approximation of $T_0$ wil be surface skin tempertaure and $e_0$ denote saturated vapor pressure when soil is not water limited. 

Note that in the above expansion, we to used $q_v =\epsilon \frac{e}{P}$ to calculate the Bowen ratio using easily measured water vapor pressure rather than the specific humidity. Some typical values for $\beta$  are:

$$ \beta \simeq 5 \;{(\rm semi\,arid)} \qquad \beta \simeq 0.5 \;{\rm (grassland)} \qquad \beta \simeq 0.2 \, {\rm (irrigated \,agriculture)} \qquad \beta \simeq 0.1\,{\rm (openwater)}$$

It is clear that the Bowen ratio decreases over moist surfaces as most energy is going to evaporation that is larger $LE\;$ and smaller $H$.  As explained, if the Bowen ratio is given and we have $Q_n$,  one can use these two equations to  estimate the sensible $H\;$ and latent $LE$ heat fluxes:

$$ \begin{cases}
Q_n = LE+H\\
H = \beta LE
\end{cases} \Rightarrow \quad LE = \frac{Q_n}{1+\beta} \quad \;{\rm and}\; \quad H= \frac{\beta Q_n}{1+\beta} . $$

**Penman method:** Up to now, we have covered both **flux-based aerodynamic** methods and an **energy balance** approach for calculating evaporative and sensible heat fluxes. The methods that combine these two, known as **combined** methods, are the best models of ET that we currently have because their solution is constrained to both flux models and land surface energy balance equation.  The most popular of which is the **Penman-Monteith** equation:

$$ LE =  \frac{\gamma^*}{\triangle + \gamma^*} \rho_a \mathcal{L}_{lv} \frac{(q_{as}(z)-q_{a}(z))}{r_{av}} + \frac{\triangle}{\triangle + \gamma^*} Q_n \quad \rm [Wm^{-2}] $$

where

- **IMPORTANT:** The unit of the above equation is based on the enegry flux $\left\lbrack {\textrm{Wm}}^{-2} \right\rbrack \ldotp$ To convert it to $\left\lbrack {\textrm{ms}}^{-1} \right\rbrack$, we need to divide it by $\mathcal{L}{\;}_{\ell v} \times \rho {\;}_w$, where $\rho {\;}_w$ is the density of water and then  
- $\triangle =\frac{{\mathcal{L}}_{\ell v} }{R_v }\times \frac{e_{sa} }{T_a^2 }=\approx \frac{e_{s0} -e_{sa} }{T_0 -T_a }~~{[\mathrm{P}\mathrm{a}\,{\mathrm{K}}^{-1} ]}$ is the slope of the CC equation. For the SI system, a simpler equation is $\Delta =\frac{\;4098\;e_{\textrm{as}} \left(z\right)}{{\left(237\ldotp 30+T_a \right)}^2 }$.  
- $\gamma =\frac{P\,c_p }{\varepsilon {\mathcal{L}}_{\ell v} }\;\textrm{and}\;\gamma^* =\gamma \cdot \frac{r_{av} }{r_{ah} }$  $\left\lbrack \textrm{Pa}\;{\mathrm{K}}^{-1} \right\rbrack$ are called psychrometric constants. As we discussed while $\gamma$ and $\gamma^*$ are theoretically different, we practically assume $\gamma =\gamma {\;}^*$.  
- $q_{\textrm{as}} \left(z\right)$ is air saturated specific humidity and $q_a \left(z\right)$ is the air specific humidity both above the surface (e.g., $z=2\;$m). 

As we can see the PPenman method model uses a weighted average of the aerodynamic method (left) and an available energy term (right) and there is no need to compuate the saturated specific humidity near soil surface. In other words, this model does not any infromation about the air temperature near the soil surface, which is a difficult variable to measure. 

____
**Example Probelm 4.3.:** Assume that the net radiation is 150 W/m2 and the ground heat flux is 25 W/m2, air temperature is 20 degrees Celsius, relative humidity of 20\%, air pressure of 101.3 kPa, and wind speed of 1.2 m/s. All measurements are at height of 2 m above the water surface. Assume roughness height $z_0$ equal to 0.02 cm. Compute the evaporation flux from a free water surface based on the combined method. 

```python
# Inputs
R_n = 150  # net radiation [W/m^2]
G = 25  # ground heat flux [W/m^2]
rho_w = 1000  # density of water [kg/m^3]
c_p = 1004  # specific heat of dry air [J/kg.K]
kappa = 0.41  # von Karman constant
RH = 0.20  # relative humidity
z = 2  # measurement elevation [m]
z_0 = 0.02 / 100  # roughness height [m]
eps = 0.622
u_2 = 1.2  # wind velocity [m/s]
P = 101.3 * 10**3  # Pressure [Pa]
T_a = 20  # air temperature [°C]
R_d = 287  # specific gas constant of dry air [J/kg.K]

# Saturated water vapor pressure function
def e_s(T_a):
    return 611 * np.exp(17.27 * T_a / (237.3 + T_a))

# Solution
e_a = RH * e_s(T_a)  # air vapor pressure [Pa]
T_v = (T_a + 273.15) / (1 - (1 - eps) * e_a / P)  # virtual air temperature [K]
rho_a = P / (R_d * T_v)  # air density [kg/m^3]
r_av = (np.log(z / z_0))**2 / (kappa**2 * u_2)  # resistance [s/m]
L_lv = (2500 - 2.36 * T_a) * 10**3  # latent heat of vaporization [J/kg]
Delta = 4098 * e_s(T_a) / (237.3 + T_a)**2  # slope of the CC equation
gamma = c_p * P / (eps * L_lv)  # psychrometric constant [Pa/°C]
E_a = rho_a * (eps / P) * (e_s(T_a) - e_a) / r_av * (1 / rho_w) * 1000 * 86400  # aerodynamic part [mm/day]
Q_n = R_n - G
E_r = Q_n / (L_lv * rho_w) * 1000 * 86400  # energy part [mm/day]
E_t = gamma / (Delta + gamma) * E_a + Delta / (Delta + gamma) * E_r  # total evaporation [mm/day]

print(f"Evaporation by the PM model is {E_t:.3f} [mm/day]")
```

**Penman-Monteith Model**

The Penman's original equation could be applied only over water, bare soil, and short vegetation as it assumes that the surface is completely saturated and does not account for the effects of vegetaiton transpiration. Therefore, one of the Penman's students, John Monteith, modified the original equation to account for the effect of the canopy and calculate ET.  As mentioned previously, the actual resistance to vapor flux is:

$$r_{\textrm{av}} :=r_c +r_{\textrm{av}}$$

where $r_c$ and $r_{av}$ are the canopy and aerodynamic resistance parameters. In other words, the total resostance to water vapor flux is sum of aerodynamic and canopy resistance. 

Now if we just update the value of $\gamma^*$ in the original Penman equation as follows:

$$\gamma^* =\gamma \cdot \frac{r_{av} }{r_{ah} }:=\gamma \cdot \frac{r_{av} +r_c }{r_{ah} }\simeq \gamma \left(1+\frac{r_c }{r_{ah} }\right)$$

with the assumption that $r_{ah} =r_{av}$, we have the Penman-Monteith equation, which has been the most widely used method for computation of ET fluxes in the past decades. Typically, for brevity, we could assume $r_{ah} =r_{av} =r_a$.  Using  this, the final form of the Penman-Monteith equation is:

$$LE=\frac{\gamma^* }{\triangle +\gamma^* }\rho_a {\mathcal{L}}_{lv} \frac{(q_{sa} -q_a )}{r_a }+\frac{\triangle }{\triangle +\gamma^* }Q_n ~~~~{\mathrm{w}\mathrm{h}\mathrm{e}\mathrm{r}\mathrm{e}}~~~~\gamma^* =\gamma \left(1+\frac{r_c }{r_a }\right)$$

Therefore, ET can be estimated if one has the required measurements, namely net radiation, air temperature, air humidity, wind speed as well as estimates for the roughness lengths and canopy resistance.

**FAO Standardized Penman-Monteith Model**

There have also been attempts to further simplify the Penman-Monteith equation for more practical use. In 1998, the Food and Agriculture Organization (FAO) of the UN put out a simplified standardized method of the equation as follows:

$${\textrm{ET}}_0 =\frac{\;0\ldotp 408\Delta \;\left(R_n -G\right)+\gamma \;\frac{\;900}{T_a +273}u_2 \left(e_{\textrm{sa}} \left(z\right)-e_a \left(z\right)\right)}{\Delta +\gamma \;\left(1+0\ldotp 34\;u_2 \right)}$$


- ${\textrm{ET}}_0$: refrence ET $\left\lbrack \textrm{mm}\;{\textrm{day}}^{-1} \right\rbrack$
- $R_n$: net radiation at the crop surface $\left\lbrack \textrm{MJ}\;{\mathrm{m}}^{-2} {\textrm{day}}^{-1} \right\rbrack$ 
- $G$: soil heat flux density $\left\lbrack \textrm{MJ}\;{\mathrm{m}}^{-2} {\textrm{day}}^{-1} \right\rbrack$
- $T_a$: mean daily air temperature at 2 m heihgt $\left\lbrack \mathrm{C}\right\rbrack$ 
- $u_2$: wind speed at 2 m height $\left\lbrack {\textrm{ms}}^{-1} \right\rbrack$ 
- $e_{\textrm{sa}}$: saturated water vapor pressure at 2 m height $\left\lbrack \textrm{kPa}\right\rbrack$ 
- $e_a$: s water vapor pressure at 2 m height $\left\lbrack \textrm{kPa}\right\rbrack$ 
- $\Delta$: slope of the CC equation at $T_a$ $\left\lbrack \textrm{kPa}\;{\mathrm{C}}^{-1} \right\rbrack$ 
- $\gamma \;$: psychrometric constant $\left\lbrack \textrm{kPa}\;{\mathrm{C}}^{-1} \right\rbrack$ 

This equation is calculating the daily ET of a standardized reference crop which is a well-watered short grass (0.12 m tall).  Additionally, the equation assumes measurements are all at 2 m height.  Then using the  one can obtain the daily crop  $ET_c$ values as follows:

$$ET_c =K_c ET_0$$

where $K_c \;$ is a crop coefficient that accounts for the difference from the [reference crop](http://www.fao.org/docrep/X0490E/x0490e0b.htm)
. Various values of $K_c$ have been recorded for differing crops under various climatic conditions and growth stages.

**Priestley-Taylor model**

If meteorological data of humidity and wind velocity are not available, an even simpler form of Penman-Monteith is the Priestley Taylor equation:

$$LE=\alpha \frac{\triangle }{\triangle +\gamma }Q_n$$

where field experiments found for well watered fields, on average $\alpha \approx 1.2-1.3$.  As you can see, this equation essentially states that the aerodynamic component of the Penman-Monteith accounts for 30 percent of the total ET in a well-watered condition.
